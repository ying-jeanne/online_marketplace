```python
# ==========================================
# 1. IMPORTS & DATA PREPARATION
# ==========================================
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer

# Load the dataset
df = pd.read_csv('Dataset_Ads.csv')

# Define the "Reward" (End-to-End Success Rate)
df['Reward'] = df['CTR'] * df['Conversion Rate']

# Map the "Arms" (Ad Types) to numerical indices
arm_names = df['Ad Type'].unique()
k_arms = len(arm_names)
arm_to_idx = {name: i for i, name in enumerate(arm_names)}
df['arm'] = df['Ad Type'].map(arm_to_idx)

# Find the winner of a traditional A/B/C/D test (Static Best)
# We calculate the global average reward for each ad type
global_means = df.groupby('arm')['Reward'].mean()
best_static_arm = global_means.idxmax()
print(f"A/B Test Winner is '{arm_names[best_static_arm]}' with average reward: {global_means.max():.4f}")

# Prepare Context Features for LinUCB
# We must scale numeric features and one-hot encode categorical ones
categorical_features = ['Gender', 'Location']
numeric_features = ['Age', 'Income']

preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numeric_features),
        ('cat', OneHotEncoder(drop='first'), categorical_features) # drop='first' avoids dummy trap
    ])

# X contains our context vectors
X_context = preprocessor.fit_transform(df)
n_features = X_context.shape[1]
print(f"Number of context features extracted: {n_features}")


# ==========================================
# 2. ALGORITHM DEFINITIONS
# ==========================================

class StaticBest:
    """Always chooses the historical best arm (mimics deploying an A/B test winner)"""
    def __init__(self, best_arm):
        self.best_arm = best_arm
        
    def select_arm(self):
        return self.best_arm
    
    def update(self, arm, reward):
        pass # Static policy does not learn online

class EpsilonGreedy:
    """Standard MAB: Explores with probability epsilon, exploits the best mean otherwise."""
    def __init__(self, n_arms, epsilon=0.1):
        self.n_arms = n_arms
        self.epsilon = epsilon
        self.counts = np.zeros(n_arms)
        self.values = np.zeros(n_arms)
        
    def select_arm(self):
        if np.random.random() < self.epsilon:
            return np.random.randint(self.n_arms) # Explore
        else:
            return np.argmax(self.values) # Exploit
            
    def update(self, arm, reward):
        self.counts[arm] += 1
        n = self.counts[arm]
        value = self.values[arm]
        # Incremental average update
        self.values[arm] = value + (reward - value) / n

class UCB1:
    """Standard MAB: Explores based on the Upper Confidence Bound."""
    def __init__(self, n_arms):
        self.n_arms = n_arms
        self.counts = np.zeros(n_arms)
        self.values = np.zeros(n_arms)
        self.total_pulls = 0
        
    def select_arm(self):
        self.total_pulls += 1
        # If any arm hasn't been pulled yet, pull it to get baseline data
        for arm in range(self.n_arms):
            if self.counts[arm] == 0:
                return arm
        
        # Calculate UCB for all arms
        ucb_values = np.zeros(self.n_arms)
        for arm in range(self.n_arms):
            bonus = np.sqrt((2 * np.log(self.total_pulls)) / self.counts[arm])
            ucb_values[arm] = self.values[arm] + bonus
        return np.argmax(ucb_values)
        
    def update(self, arm, reward):
        self.counts[arm] += 1
        n = self.counts[arm]
        value = self.values[arm]
        self.values[arm] = value + (reward - value) / n

class LinUCB:
    """Contextual MAB: Uses user features to predict rewards and calculate UCB."""
    def __init__(self, n_arms, n_features, alpha=0.5):
        self.n_arms = n_arms
        self.n_features = n_features
        self.alpha = alpha
        
        # Initialize A matrix (identity) and b vector (zeros) for each arm
        self.A = [np.eye(n_features) for _ in range(n_arms)]
        self.b = [np.zeros((n_features, 1)) for _ in range(n_arms)]
        
    def select_arm(self, context):
        context = context.reshape(-1, 1) # Ensure column vector
        p = np.zeros(self.n_arms)
        
        for arm in range(self.n_arms):
            A_inv = np.linalg.inv(self.A[arm])
            theta = A_inv @ self.b[arm]
            
            # Expected reward + Uncertainty bonus
            expected_reward = theta.T @ context
            uncertainty = self.alpha * np.sqrt(context.T @ A_inv @ context)
            
            p[arm] = expected_reward + uncertainty
            
        return np.argmax(p)
        
    def update(self, arm, reward, context):
        context = context.reshape(-1, 1)
        self.A[arm] = self.A[arm] + (context @ context.T)
        self.b[arm] = self.b[arm] + (reward * context)


# ==========================================
# 3. THE REPLAY EVALUATOR (Li et al. 2010)
# ==========================================

def run_replay_evaluation(policy, df, context_matrix, is_contextual=False):
    """
    Simulates the unbiased offline evaluation. 
    Only keeps events where the policy's chosen arm matches the historical logging policy.
    """
    # Shuffle indices to ensure streaming is random
    np.random.seed(42)
    indices = np.random.permutation(len(df))
    
    avg_reward_history = []
    cumulative_reward = 0
    matches = 0
    
    for idx in indices:
        logged_arm = df['arm'].iloc[idx]
        reward = df['Reward'].iloc[idx]
        
        if is_contextual:
            context = context_matrix[idx]
            chosen_arm = policy.select_arm(context)
        else:
            chosen_arm = policy.select_arm()
            
        # THE MATCH FILTER (Replay Logic)
        if chosen_arm == logged_arm:
            # Policy updates ONLY on a match
            if is_contextual:
                policy.update(chosen_arm, reward, context)
            else:
                policy.update(chosen_arm, reward)
                
            matches += 1
            cumulative_reward += reward
            avg_reward_history.append(cumulative_reward / matches)
            
    return avg_reward_history, matches

# ==========================================
# 4. EXECUTION & VISUALIZATION
# ==========================================

# Initialize policies
policy_ab_test = StaticBest(best_static_arm)
policy_eps_greedy = EpsilonGreedy(n_arms=k_arms, epsilon=0.1)
policy_ucb = UCB1(n_arms=k_arms)
policy_linucb = LinUCB(n_arms=k_arms, n_features=n_features, alpha=0.5)

print("\nRunning Replay Evaluator...")
hist_ab, matches_ab = run_replay_evaluation(policy_ab_test, df, X_context, is_contextual=False)
hist_eps, matches_eps = run_replay_evaluation(policy_eps_greedy, df, X_context, is_contextual=False)
hist_ucb, matches_ucb = run_replay_evaluation(policy_ucb, df, X_context, is_contextual=False)
hist_linucb, matches_linucb = run_replay_evaluation(policy_linucb, df, X_context, is_contextual=True)

# Because each policy might match a slightly different number of times, 
# we plot against the number of valid trials (matches) it experienced.
plt.figure(figsize=(14, 6))

# Plot 1: Learning Curve (Average Reward over Valid Trials)
plt.subplot(1, 2, 1)
plt.plot(hist_ab, label='A/B Test Winner (Static Best)', linestyle='--')
plt.plot(hist_eps, label='Epsilon-Greedy', alpha=0.8)
plt.plot(hist_ucb, label='UCB1', alpha=0.8)
plt.plot(hist_linucb, label='LinUCB (Contextual)', linewidth=2.5, color='red')

plt.title('Learning Curve: Average Reward over Time')
plt.xlabel('Valid Trials (Matches)')
plt.ylabel('Average Success Rate (CTR x CVR)')
plt.legend()
plt.grid(True, linestyle=':', alpha=0.7)

# Plot 2: Final Performance Comparison (Bar Chart)
plt.subplot(1, 2, 2)
final_rewards = [hist_ab[-1], hist_eps[-1], hist_ucb[-1], hist_linucb[-1]]
labels = ['A/B Test Winner', 'Epsilon-Greedy', 'UCB1', 'LinUCB']
colors = ['gray', 'blue', 'orange', 'red']

bars = plt.bar(labels, final_rewards, color=colors, alpha=0.8)
plt.title('Final Average Success Rate Comparison')
plt.ylabel('Average Success Rate')

# Add the exact numbers on top of the bars for clarity
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 0.0001, 
             f"{yval:.5f}", ha='center', va='bottom', fontweight='bold')

plt.tight_layout()
plt.show()

print("\n--- Final Results (Average Success Rate) ---")
print(f"1. A/B Test Winner (Native): {hist_ab[-1]:.5f} (over {matches_ab} matches)")
print(f"2. Epsilon-Greedy:           {hist_eps[-1]:.5f} (over {matches_eps} matches)")
print(f"3. UCB1:                     {hist_ucb[-1]:.5f} (over {matches_ucb} matches)")
print(f"4. LinUCB (Contextual):      {hist_linucb[-1]:.5f} (over {matches_linucb} matches)")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer

# ==========================================
# 1. DATA PREPARATION & FEATURE ENGINEERING
# ==========================================
df = pd.read_csv('Dataset_Ads.csv')
df['Reward'] = df['CTR'] * df['Conversion Rate']

# 1.1 Time Features: Extract Day of Week
df['Click Time'] = pd.to_datetime(df['Click Time'])
df['DayOfWeek'] = df['Click Time'].dt.dayofweek

# 1.2 Interaction Term: Age x Income
df['Age_x_Income'] = df['Age'] * df['Income']

# 1.3 Arm mapping
arm_names = df['Ad Type'].unique()
k_arms = len(arm_names)
arm_to_idx = {name: i for i, name in enumerate(arm_names)}
df['arm'] = df['Ad Type'].map(arm_to_idx)
best_static_arm = df.groupby('arm')['Reward'].mean().idxmax()

# 1.4 Contextual Features (User Demographics + Ad Characteristics)
categorical_features = ['Gender', 'Location', 'Ad Topic', 'Ad Placement']
numeric_features = ['Age', 'Income', 'DayOfWeek', 'Age_x_Income']

preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numeric_features),
        ('cat', OneHotEncoder(drop='first'), categorical_features)
    ])

X_context = preprocessor.fit_transform(df)
n_features = X_context.shape[1]

# ==========================================
# 2. POLICY DEFINITIONS
# ==========================================

class StaticBest:
    def __init__(self, best_arm): self.best_arm = best_arm
    def select_arm(self): return self.best_arm
    def update(self, arm, reward): pass

class EpsilonGreedy:
    def __init__(self, n_arms, epsilon=0.1):
        self.n_arms, self.epsilon = n_arms, epsilon
        self.counts, self.values = np.zeros(n_arms), np.zeros(n_arms)
    def select_arm(self):
        return np.random.randint(self.n_arms) if np.random.random() < self.epsilon else np.argmax(self.values)
    def update(self, arm, reward):
        self.counts[arm] += 1
        self.values[arm] += (reward - self.values[arm]) / self.counts[arm]

class UCB1:
    def __init__(self, n_arms):
        self.n_arms, self.counts, self.values, self.total_pulls = n_arms, np.zeros(n_arms), np.zeros(n_arms), 0
    def select_arm(self):
        self.total_pulls += 1
        if 0 in self.counts: return np.where(self.counts == 0)[0][0]
        return np.argmax(self.values + np.sqrt((2 * np.log(self.total_pulls)) / self.counts))
    def update(self, arm, reward):
        self.counts[arm] += 1
        self.values[arm] += (reward - self.values[arm]) / self.counts[arm]

class LinUCB:
    def __init__(self, n_arms, n_features, alpha=0.1):
        self.n_arms, self.n_features, self.alpha = n_arms, n_features, alpha
        self.A = [np.eye(n_features) for _ in range(n_arms)]
        self.b = [np.zeros((n_features, 1)) for _ in range(n_arms)]
    def select_arm(self, context):
        context = context.reshape(-1, 1)
        p = [ ( (np.linalg.inv(self.A[a]) @ self.b[a]).T @ context + 
               self.alpha * np.sqrt(context.T @ np.linalg.inv(self.A[a]) @ context) )[0,0] 
             for a in range(self.n_arms)]
        return np.argmax(p)
    def update(self, arm, reward, context):
        context = context.reshape(-1, 1)
        self.A[arm] += context @ context.T
        self.b[arm] += reward * context

def run_replay_evaluation(policy, df, context_matrix, is_contextual=False):
    np.random.seed(42)
    indices = np.random.permutation(len(df))
    avg_reward_history, cumulative_reward, matches = [], 0, 0
    for idx in indices:
        logged_arm, reward = df['arm'].iloc[idx], df['Reward'].iloc[idx]
        context = context_matrix[idx] if is_contextual else None
        chosen_arm = policy.select_arm(context) if is_contextual else policy.select_arm()
        if chosen_arm == logged_arm:
            if is_contextual: policy.update(chosen_arm, reward, context)
            else: policy.update(chosen_arm, reward)
            matches += 1
            cumulative_reward += reward
            avg_reward_history.append(cumulative_reward / matches)
    return avg_reward_history, matches

# ==========================================
# 3. EXPERIMENT EXECUTION
# ==========================================
print("Evaluating Baselines...")
hist_ab, _ = run_replay_evaluation(StaticBest(best_static_arm), df, X_context, False)
hist_eps, _ = run_replay_evaluation(EpsilonGreedy(k_arms), df, X_context, False)
hist_ucb, _ = run_replay_evaluation(UCB1(k_arms), df, X_context, False)

print("Running Alpha Sweep for LinUCB...")
alphas = [0.01, 0.1, 0.5, 1.0]
linucb_results = {a: run_replay_evaluation(LinUCB(k_arms, n_features, a), df, X_context, True) for a in alphas}

# ==========================================
# 4. FINAL RESULTS & VISUALIZATION
# ==========================================
plt.figure(figsize=(16, 7))

# Plot 1: Learning Curves
plt.subplot(1, 2, 1)
plt.plot(hist_ab, label='A/B Test Winner (Static)', linestyle='--', color='black')
plt.plot(hist_eps, label='Epsilon-Greedy', alpha=0.6)
plt.plot(hist_ucb, label='Standard UCB', alpha=0.6)
for a, (hist, _) in linucb_results.items():
    plt.plot(hist, label=f'LinUCB (alpha={a})', linewidth=2)

plt.title('Learning Curve: Success Rate Comparison')
plt.xlabel('Valid Trials (Matches)'); plt.ylabel('Avg Success Rate'); plt.legend(); plt.grid(True, alpha=0.3)

# Plot 2: Final Comparison Bar Chart
plt.subplot(1, 2, 2)
# Collect best performing alpha for the summary
best_a = max(linucb_results, key=lambda k: linucb_results[k][0][-1])
best_lin_hist, _ = linucb_results[best_a]

labels = ['A/B Winner', 'Eps-Greedy', 'UCB1', f'LinUCB (α={best_a})']
final_vals = [hist_ab[-1], hist_eps[-1], hist_ucb[-1], best_lin_hist[-1]]
colors = ['gray', 'blue', 'orange', 'red']

bars = plt.bar(labels, final_vals, color=colors, alpha=0.8)
plt.title('Final Decision-Making Performance'); plt.ylabel('Success Rate')
for bar in bars:
    plt.text(bar.get_x()+bar.get_width()/2, bar.get_height()+0.0001, f"{bar.get_height():.5f}", ha='center', fontweight='bold')

plt.tight_layout(); plt.show()

print("\n--- SUMMARY OF FINAL PERFORMANCE ---")
print(f"1. A/B Test Winner:        {hist_ab[-1]:.6f}")
print(f"2. Epsilon-Greedy:         {hist_eps[-1]:.6f}")
print(f"3. Standard UCB:           {hist_ucb[-1]:.6f}")
print(f"4. LinUCB (Best Alpha):    {best_lin_hist[-1]:.6f}")
lift = ((best_lin_hist[-1] / hist_ab[-1]) - 1) * 100
print(f"\nFinal Improvement of LinUCB over A/B Testing: {lift:.2f}%")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from tqdm.auto import tqdm  # Auto handles both terminal and notebook environments

# ==========================================
# 1. DATA PREPARATION (Avazu Dataset)
# ==========================================
df = df_sampled.copy()
df['Reward'] = df['click']
df['hour_of_day'] = df['hour'].apply(lambda x: int(str(x)[-2:]))

arm_names = df['banner_pos'].unique()
k_arms = len(arm_names)
arm_to_idx = {name: i for i, name in enumerate(arm_names)}
df['arm'] = df['banner_pos'].map(arm_to_idx)

global_means = df.groupby('arm')['Reward'].mean()
best_static_mean = global_means.max()
best_static_arm = global_means.idxmax()

categorical_features = ['C1', 'site_category', 'app_category', 'device_type', 'device_conn_type']
numeric_features = ['hour_of_day']

preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numeric_features),
        ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_features)
    ])

X_context = preprocessor.fit_transform(df)
if hasattr(X_context, "toarray"):
    X_context = X_context.toarray()
n_features = X_context.shape[1]

# ==========================================
# 2. ALGORITHM DEFINITIONS
# ==========================================

class StaticBest:
    def __init__(self, best_arm): self.best_arm = best_arm
    def select_arm(self, context=None): return self.best_arm
    def update(self, arm, reward, context=None): pass

class EpsilonGreedy:
    def __init__(self, n_arms, epsilon=0.05):
        self.n_arms, self.epsilon = n_arms, epsilon
        self.counts, self.values = np.zeros(n_arms), np.zeros(n_arms)
    def select_arm(self, context=None):
        return np.random.randint(self.n_arms) if np.random.random() < self.epsilon else np.argmax(self.values)
    def update(self, arm, reward, context=None):
        self.counts[arm] += 1
        self.values[arm] += (reward - self.values[arm]) / self.counts[arm]

class UCB1:
    def __init__(self, n_arms):
        self.n_arms, self.counts, self.values, self.total_pulls = n_arms, np.zeros(n_arms), np.zeros(n_arms), 0
    def select_arm(self, context=None):
        self.total_pulls += 1
        if 0 in self.counts: return np.where(self.counts == 0)[0][0]
        return np.argmax(self.values + np.sqrt((2 * np.log(self.total_pulls)) / self.counts))
    def update(self, arm, reward, context=None):
        self.counts[arm] += 1
        self.values[arm] += (reward - self.values[arm]) / self.counts[arm]

class LinUCB:
    def __init__(self, n_arms, n_features, alpha=0.1):
        self.n_arms, self.n_features, self.alpha = n_arms, n_features, alpha
        self.A = [np.eye(n_features) for _ in range(n_arms)]
        self.b = [np.zeros((n_features, 1)) for _ in range(n_arms)]
    def select_arm(self, context):
        context = context.reshape(-1, 1)
        p = [ ( (np.linalg.inv(self.A[a]) @ self.b[a]).T @ context + 
               self.alpha * np.sqrt(context.T @ np.linalg.inv(self.A[a]) @ context) )[0,0] 
             for a in range(self.n_arms)]
        return np.argmax(p)
    def update(self, arm, reward, context):
        context = context.reshape(-1, 1)
        self.A[arm] += context @ context.T
        self.b[arm] += reward * context

# ==========================================
# 3. REPLAY EVALUATOR (with tqdm)
# ==========================================

def run_replay_evaluation(policy, df, context_matrix, best_static_mean, is_contextual=False, desc="Evaluating"):
    np.random.seed(42)
    indices = np.random.permutation(len(df))
    avg_reward_history, regret_history, matches, cumulative_reward = [], [], 0, 0
    
    # Wrap indices with tqdm for progress tracking
    for idx in tqdm(indices, desc=desc, leave=False):
        logged_arm, reward = df['arm'].iloc[idx], df['Reward'].iloc[idx]
        context = context_matrix[idx] if is_contextual else None
        chosen_arm = policy.select_arm(context) if is_contextual else policy.select_arm()
        
        if chosen_arm == logged_arm:
            if is_contextual: policy.update(chosen_arm, reward, context)
            else: policy.update(chosen_arm, reward)
            matches += 1
            cumulative_reward += reward
            avg_reward_history.append(cumulative_reward / matches)
            regret_history.append((best_static_mean * matches) - cumulative_reward)
            
    return avg_reward_history, regret_history, matches

# ==========================================
# 4. RUN EXPERIMENTS
# ==========================================

print("Running Policy Comparisons...")
res_ab = run_replay_evaluation(StaticBest(best_static_arm), df, X_context, best_static_mean, False, desc="Static Best")
res_eps = run_replay_evaluation(EpsilonGreedy(k_arms, epsilon=0.05), df, X_context, best_static_mean, False, desc="Epsilon-Greedy")
res_ucb = run_replay_evaluation(UCB1(k_arms), df, X_context, best_static_mean, False, desc="UCB1")
res_lin = run_replay_evaluation(LinUCB(k_arms, n_features, alpha=0.1), df, X_context, best_static_mean, True, desc="LinUCB")

# ==========================================
# 5. VISUALIZATION
# ==========================================

plt.figure(figsize=(18, 6))

# Plot 1: Average CTR (Clipped to 0.3)
plt.subplot(1, 3, 1)
plt.plot(res_ab[0], label='Static Best', linestyle='--', color='black')
plt.plot(res_eps[0], label='Epsilon-Greedy')
plt.plot(res_ucb[0], label='UCB1')
plt.plot(res_lin[0], label='LinUCB', color='red', linewidth=2)
plt.title('Average CTR over Valid Trials')
plt.xlabel('Valid Trials'); plt.ylabel('Average CTR')
plt.ylim(0, 0.3) # Y-AXIS CLIPPED TO 0.3
plt.legend(); plt.grid(True, alpha=0.3)

# Plot 2: Cumulative Regret
plt.subplot(1, 3, 2)
plt.plot(res_ab[1], label='Static Best', linestyle='--', color='black')
plt.plot(res_eps[1], label='Epsilon-Greedy')
plt.plot(res_ucb[1], label='UCB1')
plt.plot(res_lin[1], label='LinUCB', color='red', linewidth=2)
plt.title('Cumulative Regret')
plt.xlabel('Valid Trials'); plt.ylabel('Regret')
plt.legend(); plt.grid(True, alpha=0.3)

# Plot 3: Final Performance Summary
plt.subplot(1, 3, 3)
final_ctrs = [res_ab[0][-1], res_eps[0][-1], res_ucb[0][-1], res_lin[0][-1]]
labels = ['Static', 'Eps-Greedy', 'UCB1', 'LinUCB']
bars = plt.bar(labels, final_ctrs, color=['gray', 'blue', 'orange', 'red'])
plt.title('Final Average CTR Comparison')
for bar in bars:
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height(), f'{bar.get_height():.4f}', ha='center', va='bottom', fontweight='bold')

plt.tight_layout(); plt.show()

# Print results
print("\n--- FINAL METRICS (AVERAGE CTR) ---")
print(f"1. Static Best:        {res_ab[0][-1]:.6f}")
print(f"2. Epsilon-Greedy:     {res_eps[0][-1]:.6f}")
print(f"3. UCB1:               {res_ucb[0][-1]:.6f}")
print(f"4. LinUCB (Context):   {res_lin[0][-1]:.6f}")
lift = ((res_lin[0][-1] / res_ab[0][-1]) - 1) * 100
print(f"\nFinal Optimization Result: LinUCB achieved a {lift:.2f}% lift relative to the static winner.")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import xgboost as xgb
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from tqdm.auto import tqdm
import os

# ==========================================
# 0. CONFIGURATION & ORACLE LOADING
# ==========================================
ORACLE_PATH = 'model_weights_depth_6.json'  # Variable for easy swapping
df = df_sampled.copy()

# FILTER: Ensure banner_pos is strictly 0 or 1 for the 2-arm simulation
df = df[df['banner_pos'].isin([0, 1])].reset_index(drop=True)

# Load the trained Oracle
if os.path.exists(ORACLE_PATH):
    oracle_model = xgb.Booster()
    oracle_model.load_model(ORACLE_PATH)
    print(f"Successfully loaded Oracle: {ORACLE_PATH}")
else:
    raise FileNotFoundError(f"Oracle weights not found at {ORACLE_PATH}. Please check the path.")

# ==========================================
# 1. DATA PREPARATION (Oracle & MAB Parity)
# ==========================================
df['Reward'] = df['click']
df['hour_of_day'] = df['hour'].apply(lambda x: int(str(x)[-2:]))

# Arms setup
arm_names = [0, 1] 
k_arms = len(arm_names)
arm_to_idx = {name: i for i, name in enumerate(arm_names)}
df['arm'] = df['banner_pos'].map(arm_to_idx)

# IDENTICAL FEATURE SUBSET (The 10 "Power Features" + engineered hour)
categorical_features = ['C1', 'site_category', 'app_category', 'device_type', 'device_conn_type', 'C14', 'C17', 'C20', 'C21']
numeric_features = ['hour_of_day']
all_features = numeric_features + categorical_features

# PRE-CALCULATE ORACLE PROBABILITIES (The "Counterfactuals")
# We generate P(Click | Arm=0) and P(Click | Arm=1) for every row
print("Generating Oracle 'What-If' probabilities...")
oracle_probs = {}

for arm in arm_names:
    temp_df = df[all_features].copy()
    temp_df['banner_pos'] = arm  # Force the arm choice
    # Apply same encoding used in Oracle training
    for col in categorical_features + ['banner_pos']:
        temp_df[col] = temp_df[col].astype('category').cat.codes
    
    dmatrix = xgb.DMatrix(temp_df[['hour_of_day', 'C1', 'banner_pos', 'site_category', 'app_category', 'device_type', 'device_conn_type', 'C14', 'C17', 'C20', 'C21']])
    oracle_probs[arm] = oracle_model.predict(dmatrix)

# MAB PREPROCESSOR (One-Hot for Linear Context)
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numeric_features),
        ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_features)
    ])

X_context = preprocessor.fit_transform(df)
if hasattr(X_context, "toarray"): X_context = X_context.toarray()
n_features = X_context.shape[1]

# ==========================================
# 2. ALGORITHM DEFINITIONS
# ==========================================

class EpsilonGreedy:
    def __init__(self, n_arms, epsilon=0.05):
        self.n_arms, self.epsilon = n_arms, epsilon
        self.counts, self.values = np.zeros(n_arms), np.zeros(n_arms)
    def select_arm(self, context=None):
        return np.random.randint(self.n_arms) if np.random.random() < self.epsilon else np.argmax(self.values)
    def update(self, arm, reward, context=None):
        self.counts[arm] += 1
        self.values[arm] += (reward - self.values[arm]) / self.counts[arm]

class UCB1:
    def __init__(self, n_arms):
        self.n_arms, self.counts, self.values, self.total_pulls = n_arms, np.zeros(n_arms), np.zeros(n_arms), 0
    def select_arm(self, context=None):
        self.total_pulls += 1
        if 0 in self.counts: return np.where(self.counts == 0)[0][0]
        return np.argmax(self.values + np.sqrt((2 * np.log(self.total_pulls)) / self.counts))
    def update(self, arm, reward, context=None):
        self.counts[arm] += 1
        self.values[arm] += (reward - self.values[arm]) / self.counts[arm]

class LinUCB:
    def __init__(self, n_arms, n_features, alpha=0.1):
        self.n_arms, self.n_features, self.alpha = n_arms, n_features, alpha
        self.A = [np.eye(n_features) for _ in range(n_arms)]
        self.b = [np.zeros((n_features, 1)) for _ in range(n_arms)]
    def select_arm(self, context):
        context = context.reshape(-1, 1)
        p = []
        for a in range(self.n_arms):
            A_inv = np.linalg.inv(self.A[a])
            theta = A_inv @ self.b[a]
            p_val = (theta.T @ context + self.alpha * np.sqrt(context.T @ A_inv @ context))[0,0]
            p.append(p_val)
        return np.argmax(p)
    def update(self, arm, reward, context):
        context = context.reshape(-1, 1)
        self.A[arm] += context @ context.T
        self.b[arm] += reward * context

# ==========================================
# 3. REPLAY EVALUATOR (Oracle-Based Regret)
# ==========================================

def run_oracle_replay(policy, df, context_matrix, oracle_probs, is_contextual=False, desc="Evaluating"):
    np.random.seed(42)
    indices = np.random.permutation(len(df))
    avg_reward_history, cumulative_regret_history, matches, total_reward, total_regret = [], [], 0, 0, 0
    
    for idx in tqdm(indices, desc=desc, leave=False):
        logged_arm, reward = df['arm'].iloc[idx], df['Reward'].iloc[idx]
        context = context_matrix[idx] if is_contextual else None
        chosen_arm = policy.select_arm(context) if is_contextual else policy.select_arm()
        
        # Match condition (Replay Method)
        if chosen_arm == logged_arm:
            if is_contextual: policy.update(chosen_arm, reward, context)
            else: policy.update(chosen_arm, reward)
            
            matches += 1
            total_reward += reward
            
            # ORACLE REGRET CALCULATION
            # P(Click) for all arms at this context
            p_arm0 = oracle_probs[0][idx]
            p_arm1 = oracle_probs[1][idx]
            
            best_possible_p = max(p_arm0, p_arm1)
            chosen_p = p_arm0 if chosen_arm == 0 else p_arm1
            
            # Instantaneous Regret = Potential Gain - Actual Potential
            total_regret += (best_possible_p - chosen_p)
            
            avg_reward_history.append(total_reward / matches)
            cumulative_regret_history.append(total_regret)
            
    return avg_reward_history, cumulative_regret_history

# ==========================================
# 4. RUN EXPERIMENTS
# ==========================================

print("Running Policy Comparisons against Oracle Benchmark...")
res_eps = run_oracle_replay(EpsilonGreedy(k_arms, epsilon=0.05), df, X_context, oracle_probs, False, desc="Epsilon-Greedy")
res_ucb = run_oracle_replay(UCB1(k_arms), df, X_context, oracle_probs, False, desc="UCB1")
res_lin = run_oracle_replay(LinUCB(k_arms, n_features, alpha=0.1), df, X_context, oracle_probs, True, desc="LinUCB")

# ==========================================
# 5. VISUALIZATION
# ==========================================

plt.figure(figsize=(18, 6))

# Plot 1: Average CTR (Modified Y-axis: 0.1 to 0.3)
plt.subplot(1, 3, 1)
plt.plot(res_eps[0], label='Epsilon-Greedy')
plt.plot(res_ucb[0], label='UCB1')
plt.plot(res_lin[0], label='LinUCB', color='red', linewidth=2)
plt.title('Average CTR over Valid Trials')
plt.xlabel('Valid Trials'); plt.ylabel('Average CTR')
plt.ylim(0.1, 0.3) 
plt.legend(); plt.grid(True, alpha=0.3)

# Plot 2: Cumulative Oracle Regret
plt.subplot(1, 3, 2)
plt.plot(res_eps[1], label='Epsilon-Greedy')
plt.plot(res_ucb[1], label='UCB1')
plt.plot(res_lin[1], label='LinUCB', color='red', linewidth=2)
plt.title('Cumulative Oracle Regret')
plt.xlabel('Valid Trials'); plt.ylabel('Regret (vs Oracle)')
plt.legend(); plt.grid(True, alpha=0.3)

# Plot 3: Summary Bar Chart (No Static)
plt.subplot(1, 3, 3)
final_ctrs = [res_eps[0][-1], res_ucb[0][-1], res_lin[0][-1]]
labels = ['Eps-Greedy', 'UCB1', 'LinUCB']
bars = plt.bar(labels, final_ctrs, color=['blue', 'orange', 'red'])
plt.title('Final Average CTR Comparison')
plt.ylim(0.1, 0.3)
for bar in bars:
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height(), f'{bar.get_height():.4f}', ha='center', va='bottom', fontweight='bold')

plt.tight_layout(); plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import xgboost as xgb
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from tqdm.auto import tqdm
import os

# ==========================================
# 0. CONFIGURATION & ORACLE LOADING
# ==========================================
ORACLE_PATH = 'model_weights_depth_6.json'  # Swap weight files here
df = df_sampled.copy()

# FILTER: Ensure banner_pos is strictly 0 or 1
df = df[df['banner_pos'].isin([0, 1])].reset_index(drop=True)

if os.path.exists(ORACLE_PATH):
    oracle_model = xgb.Booster()
    oracle_model.load_model(ORACLE_PATH)
    print(f"Successfully loaded Oracle: {ORACLE_PATH}")
else:
    raise FileNotFoundError(f"Oracle weights not found. Run the XGBoost script first.")

# ==========================================
# 1. DATA PREPARATION (Oracle & MAB)
# ==========================================
df['Reward'] = df['click']
df['hour_of_day'] = df['hour'].apply(lambda x: int(str(x)[-2:]))

arm_names = [0, 1] 
k_arms = len(arm_names)
df['arm'] = df['banner_pos']

# Features for the Oracle (Identical to training)
# categorical_features = ['C1', 'site_category', 'app_category', 'device_type', 'device_conn_type', 'C14', 'C17', 'C20', 'C21']
categorical_features = ['C1', 'site_category', 'app_category', 'device_type', 'device_conn_type']

numeric_features = ['hour_of_day']
oracle_feature_order = ['hour_of_day', 'C1', 'banner_pos', 'site_category', 'app_category', 'device_type', 'device_conn_type', 'C14', 'C17', 'C20', 'C21']

# PRE-CALCULATE ORACLE PROBABILITIES (Fast Counterfactuals)
print("Generating Oracle 'What-If' probabilities...")
oracle_probs = {}
for arm in arm_names:
    temp_df = df[oracle_feature_order].copy()
    temp_df['banner_pos'] = arm 
    for col in categorical_features + ['banner_pos']:
        temp_df[col] = temp_df[col].astype('category').cat.codes
    
    dmatrix = xgb.DMatrix(temp_df)
    oracle_probs[arm] = oracle_model.predict(dmatrix)

# MAB PREPROCESSOR
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numeric_features),
        ('cat', OneHotEncoder(handle_unknown='ignore', sparse_output=False), categorical_features)
    ])

X_context = preprocessor.fit_transform(df)
n_features = X_context.shape[1]
print(f"LinUCB Feature Dimension: {n_features}")

# ==========================================
# 2. OPTIMIZED ALGORITHM DEFINITIONS
# ==========================================

class LinUCB:
    def __init__(self, n_arms, n_features, alpha=0.1):
        self.n_arms = n_arms
        self.n_features = n_features
        self.alpha = alpha
        # MAINTAIN INVERSE DIRECTLY: O(d^2) instead of O(d^3)
        self.A_inv = [np.eye(n_features) for _ in range(n_arms)]
        self.b = [np.zeros((n_features, 1)) for _ in range(n_arms)]

    def select_arm(self, context):
        context = context.reshape(-1, 1)
        p = []
        for a in range(self.n_arms):
            # No np.linalg.inv() here!
            theta = self.A_inv[a] @ self.b[a]
            uncertainty = np.sqrt(context.T @ self.A_inv[a] @ context)
            p_val = (theta.T @ context + self.alpha * uncertainty)[0, 0]
            p.append(p_val)
        return np.argmax(p)

    def update(self, arm, reward, context):
        context = context.reshape(-1, 1)
        # Sherman-Morrison rank-1 update for A_inv
        inv = self.A_inv[arm]
        u = inv @ context
        v_T = context.T @ inv
        denominator = 1.0 + (context.T @ u)[0, 0]
        
        self.A_inv[arm] -= (u @ v_T) / denominator
        self.b[arm] += reward * context

# [Classic Definitions for EpsilonGreedy and UCB1 remained the same]
class EpsilonGreedy:
    def __init__(self, n_arms, epsilon=0.05):
        self.n_arms, self.epsilon = n_arms, epsilon
        self.counts, self.values = np.zeros(n_arms), np.zeros(n_arms)
    def select_arm(self, context=None):
        return np.random.randint(self.n_arms) if np.random.random() < self.epsilon else np.argmax(self.values)
    def update(self, arm, reward, context=None):
        self.counts[arm] += 1
        self.values[arm] += (reward - self.values[arm]) / self.counts[arm]

class UCB1:
    def __init__(self, n_arms):
        self.n_arms, self.counts, self.values, self.total_pulls = n_arms, np.zeros(n_arms), np.zeros(n_arms), 0
    def select_arm(self, context=None):
        self.total_pulls += 1
        if 0 in self.counts: return np.where(self.counts == 0)[0][0]
        return np.argmax(self.values + np.sqrt((2 * np.log(self.total_pulls)) / self.counts))
    def update(self, arm, reward, context=None):
        self.counts[arm] += 1
        self.values[arm] += (reward - self.values[arm]) / self.counts[arm]

# ==========================================
# 3. REPLAY EVALUATOR
# ==========================================

def run_oracle_replay(policy, df, context_matrix, oracle_probs, is_contextual=False, desc="Evaluating"):
    np.random.seed(42)
    indices = np.random.permutation(len(df))
    avg_reward_hist, cum_regret_hist, matches, total_reward, total_regret = [], [], 0, 0, 0
    
    for idx in tqdm(indices, desc=desc, leave=False):
        logged_arm, reward = df['arm'].iloc[idx], df['Reward'].iloc[idx]
        context = context_matrix[idx] if is_contextual else None
        chosen_arm = policy.select_arm(context) if is_contextual else policy.select_arm()
        
        if chosen_arm == logged_arm:
            if is_contextual: policy.update(chosen_arm, reward, context)
            else: policy.update(chosen_arm, reward)
            
            matches += 1
            total_reward += reward
            
            # ORACLE REGRET CALCULATION
            p0, p1 = oracle_probs[0][idx], oracle_probs[1][idx]
            best_p = max(p0, p1)
            chosen_p = p0 if chosen_arm == 0 else p1
            
            total_regret += (best_p - chosen_p)
            avg_reward_hist.append(total_reward / matches)
            cum_regret_hist.append(total_regret)
            
    return avg_reward_hist, cum_regret_hist

# ==========================================
# 4. EXECUTION & PLOTTING
# ==========================================

print("Evaluating Policies...")
res_eps = run_oracle_replay(EpsilonGreedy(k_arms), df, X_context, oracle_probs, False, "Eps-Greedy")
res_ucb = run_oracle_replay(UCB1(k_arms), df, X_context, oracle_probs, False, "UCB1")
res_lin = run_oracle_replay(LinUCB(k_arms, n_features), df, X_context, oracle_probs, True, "LinUCB")

plt.figure(figsize=(18, 6))

# Plot 1: Average CTR (0.1 to 0.3)
plt.subplot(1, 3, 1)
plt.plot(res_eps[0], label='Epsilon-Greedy')
plt.plot(res_ucb[0], label='UCB1')
plt.plot(res_lin[0], label='LinUCB', color='red', linewidth=2)
plt.title('Average CTR (Zoomed)'); plt.ylim(0.1, 0.3)
plt.xlabel('Valid Trials'); plt.ylabel('CTR'); plt.legend(); plt.grid(True, alpha=0.3)

# Plot 2: Oracle-Based Regret
plt.subplot(1, 3, 2)
plt.plot(res_eps[1], label='Epsilon-Greedy')
plt.plot(res_ucb[1], label='UCB1')
plt.plot(res_lin[1], label='LinUCB', color='red', linewidth=2)
plt.title('Cumulative Oracle Regret')
plt.xlabel('Valid Trials'); plt.ylabel('Regret vs Oracle'); plt.legend(); plt.grid(True, alpha=0.3)

# Plot 3: Final Summary (No Static)
plt.subplot(1, 3, 3)
labels, values = ['Eps-Greedy', 'UCB1', 'LinUCB'], [res_eps[0][-1], res_ucb[0][-1], res_lin[0][-1]]
bars = plt.bar(labels, values, color=['blue', 'orange', 'red'])
plt.title('Final CTR Performance'); plt.ylim(0.1, 0.3)
for b in bars: plt.text(b.get_x()+b.get_width()/2, b.get_height(), f'{b.get_height():.4f}', ha='center', va='bottom')

plt.tight_layout(); plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import xgboost as xgb
from sklearn.preprocessing import StandardScaler, OneHotEncoder, LabelEncoder
from sklearn.compose import ColumnTransformer
from tqdm.auto import tqdm
import os

# ==========================================
# 0. CONFIGURATION & ORACLE LOADING
# ==========================================
ORACLE_PATH = 'model_weights_depth_6.json'
df = df_sampled.copy()

# FILTER: Ensure banner_pos is strictly 0 or 1
df = df[df['banner_pos'].isin([0, 1])].reset_index(drop=True)

if os.path.exists(ORACLE_PATH):
    oracle_model = xgb.Booster()
    oracle_model.load_model(ORACLE_PATH)
    print(f"Successfully loaded Oracle: {ORACLE_PATH}")
else:
    raise FileNotFoundError("Oracle weight file not found.")

# ==========================================
# 1. DATA PREPARATION
# ==========================================
df['Reward'] = df['click']
df['hour_of_day'] = df['hour'].apply(lambda x: int(str(x)[-2:]))

arm_names = [0, 1] 
k_arms = len(arm_names)
df['arm'] = df['banner_pos']

# Features for Oracle (Must match original training exactly)
# categorical_features = ['C1', 'site_category', 'app_category', 'device_type', 'device_conn_type', 'C14', 'C17', 'C20', 'C21']
categorical_features = ['C1', 'site_category', 'app_category', 'device_type', 'device_conn_type']

numeric_features = ['hour_of_day']
oracle_feature_order = ['hour_of_day', 'C1', 'banner_pos', 'site_category', 'app_category', 'device_type', 'device_conn_type', 'C14', 'C17', 'C20', 'C21']

# STABLE CATEGORICAL ENCODING FOR ORACLE
# We use a fixed LabelEncoder per column to ensure 0 always means the same thing
print("Encoding features for Oracle...")
oracle_input_df = df[oracle_feature_order].copy()
for col in categorical_features + ['banner_pos']:
    le = LabelEncoder()
    oracle_input_df[col] = le.fit_transform(oracle_input_df[col].astype(str))

# PRE-CALCULATE ORACLE PROBABILITIES
print("Generating Oracle 'What-If' probabilities...")
oracle_probs = {}
for arm in arm_names:
    temp_df = oracle_input_df.copy()
    temp_df['banner_pos'] = arm  # Use the integer code for the arm
    dmatrix = xgb.DMatrix(temp_df)
    oracle_probs[arm] = oracle_model.predict(dmatrix)

# MAB PREPROCESSOR (One-Hot)
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numeric_features),
        ('cat', OneHotEncoder(handle_unknown='ignore', sparse_output=False), categorical_features)
    ])
X_context = preprocessor.fit_transform(df)
n_features = X_context.shape[1]

# ==========================================
# 2. OPTIMIZED LINUCB (Sherman-Morrison)
# ==========================================

class LinUCB:
    def __init__(self, n_arms, n_features, alpha=0.1):
        self.n_arms, self.n_features, self.alpha = n_arms, n_features, alpha
        self.A_inv = [np.eye(n_features) for _ in range(n_arms)]
        self.b = [np.zeros((n_features, 1)) for _ in range(n_arms)]

    def select_arm(self, context):
        context = context.reshape(-1, 1)
        p = []
        for a in range(self.n_arms):
            theta = self.A_inv[a] @ self.b[a]
            uncertainty = np.sqrt(context.T @ self.A_inv[a] @ context)
            p_val = (theta.T @ context + self.alpha * uncertainty)[0, 0]
            p.append(p_val)
        return np.argmax(p)

    def update(self, arm, reward, context):
        context = context.reshape(-1, 1)
        inv = self.A_inv[arm]
        u = inv @ context
        v_T = context.T @ inv
        denom = 1.0 + (context.T @ u)[0, 0]
        self.A_inv[arm] -= (u @ v_T) / denom
        self.b[arm] += reward * context

# [Classic Definitions for EpsilonGreedy and UCB1]
class EpsilonGreedy:
    def __init__(self, n_arms, epsilon=0.05):
        self.n_arms, self.epsilon = n_arms, epsilon
        self.counts, self.values = np.zeros(n_arms), np.zeros(n_arms)
    def select_arm(self, context=None):
        return np.random.randint(self.n_arms) if np.random.random() < self.epsilon else np.argmax(self.values)
    def update(self, arm, reward, context=None):
        self.counts[arm] += 1
        self.values[arm] += (reward - self.values[arm]) / self.counts[arm]

class UCB1:
    def __init__(self, n_arms):
        self.n_arms, self.counts, self.values, self.total_pulls = n_arms, np.zeros(n_arms), np.zeros(n_arms), 0
    def select_arm(self, context=None):
        self.total_pulls += 1
        if 0 in self.counts: return np.where(self.counts == 0)[0][0]
        return np.argmax(self.values + np.sqrt((2 * np.log(self.total_pulls)) / self.counts))
    def update(self, arm, reward, context=None):
        self.counts[arm] += 1
        self.values[arm] += (reward - self.values[arm]) / self.counts[arm]

# ==========================================
# 3. REPLAY EVALUATOR
# ==========================================

def run_oracle_replay(policy, df, context_matrix, oracle_probs, is_contextual=False, desc="Evaluating"):
    np.random.seed(42)
    indices = np.random.permutation(len(df))
    avg_reward_hist, cum_regret_hist, matches, total_reward, total_regret = [], [], 0, 0, 0
    
    for idx in tqdm(indices, desc=desc, leave=False):
        logged_arm, reward = df['arm'].iloc[idx], df['Reward'].iloc[idx]
        context = context_matrix[idx] if is_contextual else None
        chosen_arm = policy.select_arm(context) if is_contextual else policy.select_arm()
        
        if chosen_arm == logged_arm:
            if is_contextual: policy.update(chosen_arm, reward, context)
            else: policy.update(chosen_arm, reward)
            
            matches += 1
            total_reward += reward
            
            # ORACLE REGRET CALCULATION
            p0, p1 = oracle_probs[0][idx], oracle_probs[1][idx]
            best_p = max(p0, p1)
            chosen_p = p0 if chosen_arm == 0 else p1
            
            total_regret += (best_p - chosen_p)
            avg_reward_hist.append(total_reward / matches)
            cum_regret_hist.append(total_regret)
            
    return avg_reward_hist, cum_regret_hist, matches

# ==========================================
# 4. EXECUTION & RESULTS PRINTING
# ==========================================

print("Evaluating Policies...")
res_eps = run_oracle_replay(EpsilonGreedy(k_arms), df, X_context, oracle_probs, False, "Eps-Greedy")
res_ucb = run_oracle_replay(UCB1(k_arms), df, X_context, oracle_probs, False, "UCB1")
res_lin = run_oracle_replay(LinUCB(k_arms, n_features), df, X_context, oracle_probs, True, "LinUCB")

# PRINT ACTUAL VALUES
print("\n" + "="*40)
print(f"{'Algorithm':<15} | {'CTR':<8} | {'Regret':<10} | {'Matches':<8}")
print("-" * 45)
print(f"{'Eps-Greedy':<15} | {res_eps[0][-1]:.4f} | {res_eps[1][-1]:.2f} | {res_eps[2]}")
print(f"{'UCB1':<15} | {res_ucb[0][-1]:.4f} | {res_ucb[1][-1]:.2f} | {res_ucb[2]}")
print(f"{'LinUCB':<15} | {res_lin[0][-1]:.4f} | {res_lin[1][-1]:.2f} | {res_lin[2]}")
print("="*40)

# ==========================================
# 5. VISUALIZATION
# ==========================================
plt.figure(figsize=(18, 6))

plt.subplot(1, 3, 1)
plt.plot(res_eps[0], label='Eps-Greedy')
plt.plot(res_ucb[0], label='UCB1')
plt.plot(res_lin[0], label='LinUCB', color='red', linewidth=2)
plt.title('Average CTR (Zoomed)'); plt.ylim(0.1, 0.3)
plt.xlabel('Valid Trials'); plt.ylabel('CTR'); plt.legend(); plt.grid(True, alpha=0.3)

plt.subplot(1, 3, 2)
plt.plot(res_eps[1], label='Eps-Greedy')
plt.plot(res_ucb[1], label='UCB1')
plt.plot(res_lin[1], label='LinUCB', color='red', linewidth=2)
plt.title('Cumulative Oracle Regret')
plt.xlabel('Valid Trials'); plt.ylabel('Regret (vs Oracle)'); plt.legend(); plt.grid(True, alpha=0.3)

plt.subplot(1, 3, 3)
labels = ['Eps-Greedy', 'UCB1', 'LinUCB']
values = [res_eps[0][-1], res_ucb[0][-1], res_lin[0][-1]]
bars = plt.bar(labels, values, color=['blue', 'orange', 'red'])
plt.title('Final CTR Performance'); plt.ylim(0.1, 0.3)
for b in bars: plt.text(b.get_x()+b.get_width()/2, b.get_height(), f'{b.get_height():.4f}', ha='center', va='bottom')

plt.tight_layout(); plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import xgboost as xgb
from sklearn.preprocessing import StandardScaler, OneHotEncoder, LabelEncoder
from sklearn.compose import ColumnTransformer
from tqdm.auto import tqdm
import os

# ==========================================
# 0. CONFIGURATION & ORACLE LOADING
# ==========================================
ORACLE_PATH = 'model_weights_depth_6.json'
df = mab_eval_df.copy()

# FILTER: Ensure banner_pos is strictly 0 or 1
df = df[df['banner_pos'].isin([0, 1])].reset_index(drop=True)

if os.path.exists(ORACLE_PATH):
    oracle_model = xgb.Booster()
    oracle_model.load_model(ORACLE_PATH)
    print(f"Successfully loaded Oracle: {ORACLE_PATH}")
else:
    raise FileNotFoundError(f"Oracle weight file not found at {ORACLE_PATH}.")

# ==========================================
# 1. DATA PREPARATION (Oracle vs MAB)
# ==========================================
df['Reward'] = df['click']
df['hour_of_day'] = df['hour'].apply(lambda x: int(str(x)[-2:]))

arm_names = [0, 1] 
k_arms = len(arm_names)
df['arm'] = df['banner_pos']

mab_categorical = ['C1', 'site_category', 'app_category', 'device_type', 'device_conn_type']
mab_numeric = ['hour_of_day']

oracle_feature_order = ['hour_of_day', 'C1', 'banner_pos', 'site_category', 'app_category', 
                       'device_type', 'device_conn_type', 'C14', 'C17', 'C20', 'C21']

print("Encoding features for Oracle benchmark...")
oracle_input_df = df[oracle_feature_order].copy()
for col in oracle_feature_order:
    if col != 'hour_of_day':
        le = LabelEncoder()
        oracle_input_df[col] = le.fit_transform(oracle_input_df[col].astype(str))

print("Generating Oracle 'What-If' probabilities...")
oracle_probs = {}
for arm in arm_names:
    temp_df = oracle_input_df.copy()
    temp_df['banner_pos'] = arm  
    dmatrix = xgb.DMatrix(temp_df)
    oracle_probs[arm] = oracle_model.predict(dmatrix)

preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), mab_numeric),
        ('cat', OneHotEncoder(handle_unknown='ignore', sparse_output=False), mab_categorical)
    ])
X_context = preprocessor.fit_transform(df)
n_features = X_context.shape[1]
print(f"MAB Linear Feature Space: {n_features} dimensions")

# ==========================================
# 2. OPTIMIZED ALGORITHM DEFINITIONS
# ==========================================

class LinUCB:
    def __init__(self, n_arms, n_features, alpha=0.1):
        self.n_arms, self.n_features, self.alpha = n_arms, n_features, alpha
        self.A_inv = [np.eye(n_features) for _ in range(n_arms)]
        self.b = [np.zeros((n_features, 1)) for _ in range(n_arms)]

    def select_arm(self, context):
        context = context.reshape(-1, 1)
        p = []
        for a in range(self.n_arms):
            theta = self.A_inv[a] @ self.b[a]
            uncertainty = np.sqrt(context.T @ self.A_inv[a] @ context)
            p_val = (theta.T @ context + self.alpha * uncertainty)[0, 0]
            p.append(p_val)
        return np.argmax(p)

    def update(self, arm, reward, context):
        context = context.reshape(-1, 1)
        inv = self.A_inv[arm]
        u = inv @ context
        v_T = context.T @ inv
        denom = 1.0 + (context.T @ u)[0, 0]
        self.A_inv[arm] -= (u @ v_T) / denom
        self.b[arm] += reward * context

class EpsilonGreedy:
    def __init__(self, n_arms, epsilon=0.05):
        self.n_arms, self.epsilon = n_arms, epsilon
        self.counts, self.values = np.zeros(n_arms), np.zeros(n_arms)
    def select_arm(self, context=None):
        return np.random.randint(self.n_arms) if np.random.random() < self.epsilon else np.argmax(self.values)
    def update(self, arm, reward, context=None):
        self.counts[arm] += 1
        self.values[arm] += (reward - self.values[arm]) / self.counts[arm]

class UCB1:
    def __init__(self, n_arms):
        self.n_arms, self.counts, self.values, self.total_pulls = n_arms, np.zeros(n_arms), np.zeros(n_arms), 0
    def select_arm(self, context=None):
        self.total_pulls += 1
        if 0 in self.counts: return np.where(self.counts == 0)[0][0]
        return np.argmax(self.values + np.sqrt((2 * np.log(self.total_pulls)) / self.counts))
    def update(self, arm, reward, context=None):
        self.counts[arm] += 1
        self.values[arm] += (reward - self.values[arm]) / self.counts[arm]

# ==========================================
# 3. REPLAY EVALUATOR
# ==========================================

def run_oracle_replay(policy, df, context_matrix, oracle_probs, is_contextual=False, desc="Evaluating"):
    np.random.seed(42)
    indices = np.random.permutation(len(df))
    avg_reward_hist, cum_regret_hist, matches, total_reward, total_regret = [], [], 0, 0, 0
    
    for idx in tqdm(indices, desc=desc, leave=False):
        logged_arm, reward = df['arm'].iloc[idx], df['Reward'].iloc[idx]
        context = context_matrix[idx] if is_contextual else None
        chosen_arm = policy.select_arm(context) if is_contextual else policy.select_arm()
        
        if chosen_arm == logged_arm:
            if is_contextual: policy.update(chosen_arm, reward, context)
            else: policy.update(chosen_arm, reward)
            
            matches += 1
            total_reward += reward
            
            p0, p1 = oracle_probs[0][idx], oracle_probs[1][idx]
            best_p = max(p0, p1)
            chosen_p = p0 if chosen_arm == 0 else p1
            
            total_regret += (best_p - chosen_p)
            avg_reward_hist.append(total_reward / matches)
            cum_regret_hist.append(total_regret)
            
    return avg_reward_hist, cum_regret_hist, matches

# ==========================================
# 4. STEP 1: ALPHA HYPERPARAMETER SEARCH
# ==========================================
alphas = [0.1, 0.5, 1.0, 1.5, 2.0]
alpha_results = {}

print("Step 1: Evaluating different Alphas for LinUCB...")
for a in alphas:
    res = run_oracle_replay(LinUCB(k_arms, n_features, alpha=a), df, X_context, oracle_probs, True, f"Alpha {a}")
    alpha_results[a] = res

# Find best alpha based on final CTR
best_alpha = max(alphas, key=lambda a: alpha_results[a][0][-1])

print("\n" + "="*45)
print(f"{'Alpha':<15} | {'Final CTR':<10} | {'Total Regret':<12}")
print("-" * 45)
for a in alphas:
    print(f"{a:<15} | {alpha_results[a][0][-1]:.5f} | {alpha_results[a][1][-1]:.2f}")
print("="*45)
print(f"Best Alpha found: {best_alpha}")

# ==========================================
# 5. STEP 2: FULL COMPARISON (using Best Alpha)
# ==========================================
print("\nStep 2: Evaluating Baselines and Final Comparison...")
res_eps = run_oracle_replay(EpsilonGreedy(k_arms), df, X_context, oracle_probs, False, "Eps-Greedy")
res_ucb = run_oracle_replay(UCB1(k_arms), df, X_context, oracle_probs, False, "UCB1")
res_lin_best = alpha_results[best_alpha]

print("\n" + "="*45)
print(f"{'Algorithm':<15} | {'Final CTR':<10} | {'Total Regret':<12}")
print("-" * 45)
print(f"{'Eps-Greedy':<15} | {res_eps[0][-1]:.5f} | {res_eps[1][-1]:.2f}")
print(f"{'UCB1':<15} | {res_ucb[0][-1]:.5f} | {res_ucb[1][-1]:.2f}")
print(f"{f'LinUCB (a={best_alpha})':<15} | {res_lin_best[0][-1]:.5f} | {res_lin_best[1][-1]:.2f}")
print("="*45)

# ==========================================
# 6. VISUALIZATION
# ==========================================
plt.figure(figsize=(20, 12))

# Subplot 1: Alpha Sweep CTR
plt.subplot(2, 3, 1)
for a in alphas:
    plt.plot(alpha_results[a][0], label=f'a={a}')
plt.title('Alpha Search: CTR Progression'); plt.ylim(0.1, 0.3)
plt.xlabel('Matches'); plt.ylabel('CTR'); plt.legend(); plt.grid(True, alpha=0.3)

# Subplot 2: Alpha Sweep Regret
plt.subplot(2, 3, 2)
for a in alphas:
    plt.plot(alpha_results[a][1], label=f'a={a}')
plt.title('Alpha Search: Cumulative Regret'); plt.xlabel('Matches'); plt.ylabel('Regret'); plt.legend(); plt.grid(True, alpha=0.3)

# Subplot 3: Final Comparison CTR
plt.subplot(2, 3, 4)
plt.plot(res_eps[0], label='Eps-Greedy')
plt.plot(res_ucb[0], label='UCB1')
plt.plot(res_lin_best[0], label=f'LinUCB (a={best_alpha})', color='red', linewidth=2)
plt.title('Final: CTR Progression'); plt.ylim(0.1, 0.3)
plt.xlabel('Matches'); plt.ylabel('CTR'); plt.legend(); plt.grid(True, alpha=0.3)

# Subplot 4: Final Comparison Regret
plt.subplot(2, 3, 5)
plt.plot(res_eps[1], label='Eps-Greedy')
plt.plot(res_ucb[1], label='UCB1')
plt.plot(res_lin_best[1], label=f'LinUCB (a={best_alpha})', color='red', linewidth=2)
plt.title('Final: Cumulative Regret'); plt.xlabel('Matches'); plt.ylabel('Regret'); plt.legend(); plt.grid(True, alpha=0.3)

# Subplot 5: Final Comparison Bar Chart
plt.subplot(2, 3, 6)
labels = ['Eps-Greedy', 'UCB1', f'LinUCB (a={best_alpha})']
values = [res_eps[0][-1], res_ucb[0][-1], res_lin_best[0][-1]]
bars = plt.bar(labels, values, color=['blue', 'orange', 'red'])
plt.title('Final Performance comparison'); plt.ylim(0.1, 0.3)
for b in bars: plt.text(b.get_x()+b.get_width()/2, b.get_height(), f'{b.get_height():.4f}', ha='center', va='bottom')

plt.tight_layout(); plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import xgboost as xgb
from sklearn.preprocessing import StandardScaler, OneHotEncoder, LabelEncoder
from sklearn.compose import ColumnTransformer
from tqdm.auto import tqdm
import os

# ==========================================
# 0. CONFIGURATION & ORACLE LOADING
# ==========================================
ORACLE_PATH = 'model_weights_depth_6.json'
df = mab_eval_df.copy()

# FILTER: Ensure banner_pos is strictly 0 or 1
df = df[df['banner_pos'].isin([0, 1])].reset_index(drop=True)

if os.path.exists(ORACLE_PATH):
    oracle_model = xgb.Booster()
    oracle_model.load_model(ORACLE_PATH)
    print(f"Successfully loaded Oracle: {ORACLE_PATH}")
else:
    raise FileNotFoundError(f"Oracle weight file not found at {ORACLE_PATH}.")

# ==========================================
# 1. DATA PREPARATION (Oracle vs MAB)
# ==========================================
df['Reward'] = df['click']
df['hour_of_day'] = df['hour'].apply(lambda x: int(str(x)[-2:]))

arm_names = [0, 1] 
k_arms = len(arm_names)
df['arm'] = df['banner_pos']

mab_categorical = ['C1', 'site_category', 'app_category', 'device_type', 'device_conn_type','C14','C17']
mab_numeric = ['hour_of_day']

oracle_feature_order = ['hour_of_day', 'C1', 'banner_pos', 'site_category', 'app_category', 
                       'device_type', 'device_conn_type', 'C14', 'C17', 'C20', 'C21']

print("Encoding features for Oracle benchmark...")
oracle_input_df = df[oracle_feature_order].copy()
for col in oracle_feature_order:
    if col != 'hour_of_day':
        le = LabelEncoder()
        oracle_input_df[col] = le.fit_transform(oracle_input_df[col].astype(str))

print("Generating Oracle 'What-If' probabilities...")
oracle_probs = {}
for arm in arm_names:
    temp_df = oracle_input_df.copy()
    temp_df['banner_pos'] = arm  
    dmatrix = xgb.DMatrix(temp_df)
    oracle_probs[arm] = oracle_model.predict(dmatrix)

preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), mab_numeric),
        ('cat', OneHotEncoder(handle_unknown='ignore', sparse_output=False), mab_categorical)
    ])
X_context = preprocessor.fit_transform(df)
n_features = X_context.shape[1]
print(f"MAB Linear Feature Space: {n_features} dimensions")

# ==========================================
# 2. OPTIMIZED ALGORITHM DEFINITIONS
# ==========================================

class LinUCB:
    def __init__(self, n_arms, n_features, alpha=0.1):
        self.n_arms, self.n_features, self.alpha = n_arms, n_features, alpha
        self.A_inv = [np.eye(n_features) for _ in range(n_arms)]
        self.b = [np.zeros((n_features, 1)) for _ in range(n_arms)]

    def select_arm(self, context):
        context = context.reshape(-1, 1)
        p = []
        for a in range(self.n_arms):
            theta = self.A_inv[a] @ self.b[a]
            uncertainty = np.sqrt(context.T @ self.A_inv[a] @ context)
            p_val = (theta.T @ context + self.alpha * uncertainty)[0, 0]
            p.append(p_val)
        return np.argmax(p)

    def update(self, arm, reward, context):
        context = context.reshape(-1, 1)
        inv = self.A_inv[arm]
        u = inv @ context
        v_T = context.T @ inv
        denom = 1.0 + (context.T @ u)[0, 0]
        self.A_inv[arm] -= (u @ v_T) / denom
        self.b[arm] += reward * context

class EpsilonGreedy:
    def __init__(self, n_arms, epsilon=0.05):
        self.n_arms, self.epsilon = n_arms, epsilon
        self.counts, self.values = np.zeros(n_arms), np.zeros(n_arms)
    def select_arm(self, context=None):
        return np.random.randint(self.n_arms) if np.random.random() < self.epsilon else np.argmax(self.values)
    def update(self, arm, reward, context=None):
        self.counts[arm] += 1
        self.values[arm] += (reward - self.values[arm]) / self.counts[arm]

class UCB1:
    def __init__(self, n_arms):
        self.n_arms, self.counts, self.values, self.total_pulls = n_arms, np.zeros(n_arms), np.zeros(n_arms), 0
    def select_arm(self, context=None):
        self.total_pulls += 1
        if 0 in self.counts: return np.where(self.counts == 0)[0][0]
        return np.argmax(self.values + np.sqrt((2 * np.log(self.total_pulls)) / self.counts))
    def update(self, arm, reward, context=None):
        self.counts[arm] += 1
        self.values[arm] += (reward - self.values[arm]) / self.counts[arm]

# ==========================================
# 3. REPLAY EVALUATOR
# ==========================================

def run_oracle_replay(policy, df, context_matrix, oracle_probs, is_contextual=False, desc="Evaluating"):
    np.random.seed(42)
    indices = np.random.permutation(len(df))
    avg_reward_hist, cum_regret_hist, matches, total_reward, total_regret = [], [], 0, 0, 0
    
    for idx in tqdm(indices, desc=desc, leave=False):
        logged_arm, reward = df['arm'].iloc[idx], df['Reward'].iloc[idx]
        context = context_matrix[idx] if is_contextual else None
        chosen_arm = policy.select_arm(context) if is_contextual else policy.select_arm()
        
        if chosen_arm == logged_arm:
            if is_contextual: policy.update(chosen_arm, reward, context)
            else: policy.update(chosen_arm, reward)
            
            matches += 1
            total_reward += reward
            
            p0, p1 = oracle_probs[0][idx], oracle_probs[1][idx]
            best_p = max(p0, p1)
            chosen_p = p0 if chosen_arm == 0 else p1
            
            total_regret += (best_p - chosen_p)
            avg_reward_hist.append(total_reward / matches)
            cum_regret_hist.append(total_regret)
            
    return avg_reward_hist, cum_regret_hist, matches

# ==========================================
# 4. STEP 1: ALPHA HYPERPARAMETER SEARCH
# ==========================================
alphas = [0.1, 0.5, 1.0, 1.5, 2.0]
alpha_results = {}

print("Step 1: Evaluating different Alphas for LinUCB...")
for a in alphas:
    res = run_oracle_replay(LinUCB(k_arms, n_features, alpha=a), df, X_context, oracle_probs, True, f"Alpha {a}")
    alpha_results[a] = res

# Find best alpha based on final CTR
best_alpha = max(alphas, key=lambda a: alpha_results[a][0][-1])

print("\n" + "="*45)
print(f"{'Alpha':<15} | {'Final CTR':<10} | {'Total Regret':<12}")
print("-" * 45)
for a in alphas:
    print(f"{a:<15} | {alpha_results[a][0][-1]:.5f} | {alpha_results[a][1][-1]:.2f}")
print("="*45)
print(f"Best Alpha found: {best_alpha}")

# ==========================================
# 5. STEP 2: FULL COMPARISON (using Best Alpha)
# ==========================================
print("\nStep 2: Evaluating Baselines and Final Comparison...")
res_eps = run_oracle_replay(EpsilonGreedy(k_arms), df, X_context, oracle_probs, False, "Eps-Greedy")
res_ucb = run_oracle_replay(UCB1(k_arms), df, X_context, oracle_probs, False, "UCB1")
res_lin_best = alpha_results[best_alpha]

print("\n" + "="*45)
print(f"{'Algorithm':<15} | {'Final CTR':<10} | {'Total Regret':<12}")
print("-" * 45)
print(f"{'Eps-Greedy':<15} | {res_eps[0][-1]:.5f} | {res_eps[1][-1]:.2f}")
print(f"{'UCB1':<15} | {res_ucb[0][-1]:.5f} | {res_ucb[1][-1]:.2f}")
print(f"{f'LinUCB (a={best_alpha})':<15} | {res_lin_best[0][-1]:.5f} | {res_lin_best[1][-1]:.2f}")
print("="*45)

# ==========================================
# 6. VISUALIZATION
# ==========================================
plt.figure(figsize=(20, 12))

# Subplot 1: Alpha Sweep CTR
plt.subplot(2, 3, 1)
for a in alphas:
    plt.plot(alpha_results[a][0], label=f'a={a}')
plt.title('Alpha Search: CTR Progression'); plt.ylim(0.1, 0.3)
plt.xlabel('Matches'); plt.ylabel('CTR'); plt.legend(); plt.grid(True, alpha=0.3)

# Subplot 2: Alpha Sweep Regret
plt.subplot(2, 3, 2)
for a in alphas:
    plt.plot(alpha_results[a][1], label=f'a={a}')
plt.title('Alpha Search: Cumulative Regret'); plt.xlabel('Matches'); plt.ylabel('Regret'); plt.legend(); plt.grid(True, alpha=0.3)

# Subplot 3: Final Comparison CTR
plt.subplot(2, 3, 4)
plt.plot(res_eps[0], label='Eps-Greedy')
plt.plot(res_ucb[0], label='UCB1')
plt.plot(res_lin_best[0], label=f'LinUCB (a={best_alpha})', color='red', linewidth=2)
plt.title('Final: CTR Progression'); plt.ylim(0.1, 0.3)
plt.xlabel('Matches'); plt.ylabel('CTR'); plt.legend(); plt.grid(True, alpha=0.3)

# Subplot 4: Final Comparison Regret
plt.subplot(2, 3, 5)
plt.plot(res_eps[1], label='Eps-Greedy')
plt.plot(res_ucb[1], label='UCB1')
plt.plot(res_lin_best[1], label=f'LinUCB (a={best_alpha})', color='red', linewidth=2)
plt.title('Final: Cumulative Regret'); plt.xlabel('Matches'); plt.ylabel('Regret'); plt.legend(); plt.grid(True, alpha=0.3)

# Subplot 5: Final Comparison Bar Chart
plt.subplot(2, 3, 6)
labels = ['Eps-Greedy', 'UCB1', f'LinUCB (a={best_alpha})']
values = [res_eps[0][-1], res_ucb[0][-1], res_lin_best[0][-1]]
bars = plt.bar(labels, values, color=['blue', 'orange', 'red'])
plt.title('Final Performance comparison'); plt.ylim(0.1, 0.3)
for b in bars: plt.text(b.get_x()+b.get_width()/2, b.get_height(), f'{b.get_height():.4f}', ha='center', va='bottom')

plt.tight_layout(); plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import xgboost as xgb
from sklearn.preprocessing import StandardScaler, OneHotEncoder, LabelEncoder
from sklearn.compose import ColumnTransformer
from tqdm.auto import tqdm
import os

# ==========================================
# 0. CONFIGURATION & ORACLE LOADING
# ==========================================
ORACLE_PATH = 'model_weights_depth_6.json'
df = mab_eval_df.copy()

# FILTER: Ensure banner_pos is strictly 0 or 1
df = df[df['banner_pos'].isin([0, 1])].reset_index(drop=True)

if os.path.exists(ORACLE_PATH):
    oracle_model = xgb.Booster()
    oracle_model.load_model(ORACLE_PATH)
    print(f"Successfully loaded Oracle: {ORACLE_PATH}")
else:
    raise FileNotFoundError(f"Oracle weight file not found at {ORACLE_PATH}.")

# ==========================================
# 1. DATA PREPARATION (Oracle vs MAB)
# ==========================================
df['Reward'] = df['click']
df['hour_of_day'] = df['hour'].apply(lambda x: int(str(x)[-2:]))

arm_names = [0, 1] 
k_arms = len(arm_names)
df['arm'] = df['banner_pos']

mab_categorical = ['C1', 'site_category', 'app_category', 'device_type', 'device_conn_type','C20']
mab_numeric = ['hour_of_day']

oracle_feature_order = ['hour_of_day', 'C1', 'banner_pos', 'site_category', 'app_category', 
                       'device_type', 'device_conn_type', 'C14', 'C17', 'C20', 'C21']

print("Encoding features for Oracle benchmark...")
oracle_input_df = df[oracle_feature_order].copy()
for col in oracle_feature_order:
    if col != 'hour_of_day':
        le = LabelEncoder()
        oracle_input_df[col] = le.fit_transform(oracle_input_df[col].astype(str))

print("Generating Oracle 'What-If' probabilities...")
oracle_probs = {}
for arm in arm_names:
    temp_df = oracle_input_df.copy()
    temp_df['banner_pos'] = arm  
    dmatrix = xgb.DMatrix(temp_df)
    oracle_probs[arm] = oracle_model.predict(dmatrix)

preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), mab_numeric),
        ('cat', OneHotEncoder(handle_unknown='ignore', sparse_output=False), mab_categorical)
    ])
X_context = preprocessor.fit_transform(df)
n_features = X_context.shape[1]
print(f"MAB Linear Feature Space: {n_features} dimensions")

# ==========================================
# 2. OPTIMIZED ALGORITHM DEFINITIONS
# ==========================================

class LinUCB:
    def __init__(self, n_arms, n_features, alpha=0.1):
        self.n_arms, self.n_features, self.alpha = n_arms, n_features, alpha
        self.A_inv = [np.eye(n_features) for _ in range(n_arms)]
        self.b = [np.zeros((n_features, 1)) for _ in range(n_arms)]

    def select_arm(self, context):
        context = context.reshape(-1, 1)
        p = []
        for a in range(self.n_arms):
            theta = self.A_inv[a] @ self.b[a]
            uncertainty = np.sqrt(context.T @ self.A_inv[a] @ context)
            p_val = (theta.T @ context + self.alpha * uncertainty)[0, 0]
            p.append(p_val)
        return np.argmax(p)

    def update(self, arm, reward, context):
        context = context.reshape(-1, 1)
        inv = self.A_inv[arm]
        u = inv @ context
        v_T = context.T @ inv
        denom = 1.0 + (context.T @ u)[0, 0]
        self.A_inv[arm] -= (u @ v_T) / denom
        self.b[arm] += reward * context

class EpsilonGreedy:
    def __init__(self, n_arms, epsilon=0.05):
        self.n_arms, self.epsilon = n_arms, epsilon
        self.counts, self.values = np.zeros(n_arms), np.zeros(n_arms)
    def select_arm(self, context=None):
        return np.random.randint(self.n_arms) if np.random.random() < self.epsilon else np.argmax(self.values)
    def update(self, arm, reward, context=None):
        self.counts[arm] += 1
        self.values[arm] += (reward - self.values[arm]) / self.counts[arm]

class UCB1:
    def __init__(self, n_arms):
        self.n_arms, self.counts, self.values, self.total_pulls = n_arms, np.zeros(n_arms), np.zeros(n_arms), 0
    def select_arm(self, context=None):
        self.total_pulls += 1
        if 0 in self.counts: return np.where(self.counts == 0)[0][0]
        return np.argmax(self.values + np.sqrt((2 * np.log(self.total_pulls)) / self.counts))
    def update(self, arm, reward, context=None):
        self.counts[arm] += 1
        self.values[arm] += (reward - self.values[arm]) / self.counts[arm]

# ==========================================
# 3. REPLAY EVALUATOR
# ==========================================

def run_oracle_replay(policy, df, context_matrix, oracle_probs, is_contextual=False, desc="Evaluating"):
    np.random.seed(42)
    indices = np.random.permutation(len(df))
    avg_reward_hist, cum_regret_hist, matches, total_reward, total_regret = [], [], 0, 0, 0
    
    for idx in tqdm(indices, desc=desc, leave=False):
        logged_arm, reward = df['arm'].iloc[idx], df['Reward'].iloc[idx]
        context = context_matrix[idx] if is_contextual else None
        chosen_arm = policy.select_arm(context) if is_contextual else policy.select_arm()
        
        if chosen_arm == logged_arm:
            if is_contextual: policy.update(chosen_arm, reward, context)
            else: policy.update(chosen_arm, reward)
            
            matches += 1
            total_reward += reward
            
            p0, p1 = oracle_probs[0][idx], oracle_probs[1][idx]
            best_p = max(p0, p1)
            chosen_p = p0 if chosen_arm == 0 else p1
            
            total_regret += (best_p - chosen_p)
            avg_reward_hist.append(total_reward / matches)
            cum_regret_hist.append(total_regret)
            
    return avg_reward_hist, cum_regret_hist, matches

# ==========================================
# 4. STEP 1: ALPHA HYPERPARAMETER SEARCH
# ==========================================
alphas = [0.1, 0.5, 1.0, 1.5, 2.0]
alpha_results = {}

print("Step 1: Evaluating different Alphas for LinUCB...")
for a in alphas:
    res = run_oracle_replay(LinUCB(k_arms, n_features, alpha=a), df, X_context, oracle_probs, True, f"Alpha {a}")
    alpha_results[a] = res

# Find best alpha based on final CTR
best_alpha = max(alphas, key=lambda a: alpha_results[a][0][-1])

print("\n" + "="*45)
print(f"{'Alpha':<15} | {'Final CTR':<10} | {'Total Regret':<12}")
print("-" * 45)
for a in alphas:
    print(f"{a:<15} | {alpha_results[a][0][-1]:.5f} | {alpha_results[a][1][-1]:.2f}")
print("="*45)
print(f"Best Alpha found: {best_alpha}")

# ==========================================
# 5. STEP 2: FULL COMPARISON (using Best Alpha)
# ==========================================
print("\nStep 2: Evaluating Baselines and Final Comparison...")
res_eps = run_oracle_replay(EpsilonGreedy(k_arms), df, X_context, oracle_probs, False, "Eps-Greedy")
res_ucb = run_oracle_replay(UCB1(k_arms), df, X_context, oracle_probs, False, "UCB1")
res_lin_best = alpha_results[best_alpha]

print("\n" + "="*45)
print(f"{'Algorithm':<15} | {'Final CTR':<10} | {'Total Regret':<12}")
print("-" * 45)
print(f"{'Eps-Greedy':<15} | {res_eps[0][-1]:.5f} | {res_eps[1][-1]:.2f}")
print(f"{'UCB1':<15} | {res_ucb[0][-1]:.5f} | {res_ucb[1][-1]:.2f}")
print(f"{f'LinUCB (a={best_alpha})':<15} | {res_lin_best[0][-1]:.5f} | {res_lin_best[1][-1]:.2f}")
print("="*45)

# ==========================================
# 6. VISUALIZATION
# ==========================================
plt.figure(figsize=(20, 12))

# Subplot 1: Alpha Sweep CTR
plt.subplot(2, 3, 1)
for a in alphas:
    plt.plot(alpha_results[a][0], label=f'a={a}')
plt.title('Alpha Search: CTR Progression'); plt.ylim(0.1, 0.3)
plt.xlabel('Matches'); plt.ylabel('CTR'); plt.legend(); plt.grid(True, alpha=0.3)

# Subplot 2: Alpha Sweep Regret
plt.subplot(2, 3, 2)
for a in alphas:
    plt.plot(alpha_results[a][1], label=f'a={a}')
plt.title('Alpha Search: Cumulative Regret'); plt.xlabel('Matches'); plt.ylabel('Regret'); plt.legend(); plt.grid(True, alpha=0.3)

# Subplot 3: Final Comparison CTR
plt.subplot(2, 3, 4)
plt.plot(res_eps[0], label='Eps-Greedy')
plt.plot(res_ucb[0], label='UCB1')
plt.plot(res_lin_best[0], label=f'LinUCB (a={best_alpha})', color='red', linewidth=2)
plt.title('Final: CTR Progression'); plt.ylim(0.1, 0.3)
plt.xlabel('Matches'); plt.ylabel('CTR'); plt.legend(); plt.grid(True, alpha=0.3)

# Subplot 4: Final Comparison Regret
plt.subplot(2, 3, 5)
plt.plot(res_eps[1], label='Eps-Greedy')
plt.plot(res_ucb[1], label='UCB1')
plt.plot(res_lin_best[1], label=f'LinUCB (a={best_alpha})', color='red', linewidth=2)
plt.title('Final: Cumulative Regret'); plt.xlabel('Matches'); plt.ylabel('Regret'); plt.legend(); plt.grid(True, alpha=0.3)

# Subplot 5: Final Comparison Bar Chart
plt.subplot(2, 3, 6)
labels = ['Eps-Greedy', 'UCB1', f'LinUCB (a={best_alpha})']
values = [res_eps[0][-1], res_ucb[0][-1], res_lin_best[0][-1]]
bars = plt.bar(labels, values, color=['blue', 'orange', 'red'])
plt.title('Final Performance comparison'); plt.ylim(0.1, 0.3)
for b in bars: plt.text(b.get_x()+b.get_width()/2, b.get_height(), f'{b.get_height():.4f}', ha='center', va='bottom')

plt.tight_layout(); plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import xgboost as xgb
from sklearn.preprocessing import StandardScaler, OneHotEncoder, LabelEncoder
from sklearn.compose import ColumnTransformer
from tqdm.auto import tqdm
import os

# ==========================================
# 0. CONFIGURATION & ORACLE LOADING
# ==========================================
ORACLE_PATH = 'model_weights_depth_6.json'
df = mab_eval_df.copy()

# FILTER: Ensure banner_pos is strictly 0 or 1
df = df[df['banner_pos'].isin([0, 1])].reset_index(drop=True)

if os.path.exists(ORACLE_PATH):
    oracle_model = xgb.Booster()
    oracle_model.load_model(ORACLE_PATH)
    print(f"Successfully loaded Oracle: {ORACLE_PATH}")
else:
    raise FileNotFoundError(f"Oracle weight file not found at {ORACLE_PATH}.")

# ==========================================
# 1. DATA PREPARATION (Oracle vs MAB)
# ==========================================
df['Reward'] = df['click']
df['hour_of_day'] = df['hour'].apply(lambda x: int(str(x)[-2:]))

arm_names = [0, 1] 
k_arms = len(arm_names)
df['arm'] = df['banner_pos']

mab_categorical = ['C1', 'site_category', 'app_category', 'device_type', 'device_conn_type','C21']
mab_numeric = ['hour_of_day']

oracle_feature_order = ['hour_of_day', 'C1', 'banner_pos', 'site_category', 'app_category', 
                       'device_type', 'device_conn_type', 'C14', 'C17', 'C20', 'C21']

print("Encoding features for Oracle benchmark...")
oracle_input_df = df[oracle_feature_order].copy()
for col in oracle_feature_order:
    if col != 'hour_of_day':
        le = LabelEncoder()
        oracle_input_df[col] = le.fit_transform(oracle_input_df[col].astype(str))

print("Generating Oracle 'What-If' probabilities...")
oracle_probs = {}
for arm in arm_names:
    temp_df = oracle_input_df.copy()
    temp_df['banner_pos'] = arm  
    dmatrix = xgb.DMatrix(temp_df)
    oracle_probs[arm] = oracle_model.predict(dmatrix)

preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), mab_numeric),
        ('cat', OneHotEncoder(handle_unknown='ignore', sparse_output=False), mab_categorical)
    ])
X_context = preprocessor.fit_transform(df)
n_features = X_context.shape[1]
print(f"MAB Linear Feature Space: {n_features} dimensions")

# ==========================================
# 2. OPTIMIZED ALGORITHM DEFINITIONS
# ==========================================

class LinUCB:
    def __init__(self, n_arms, n_features, alpha=0.1):
        self.n_arms, self.n_features, self.alpha = n_arms, n_features, alpha
        self.A_inv = [np.eye(n_features) for _ in range(n_arms)]
        self.b = [np.zeros((n_features, 1)) for _ in range(n_arms)]

    def select_arm(self, context):
        context = context.reshape(-1, 1)
        p = []
        for a in range(self.n_arms):
            theta = self.A_inv[a] @ self.b[a]
            uncertainty = np.sqrt(context.T @ self.A_inv[a] @ context)
            p_val = (theta.T @ context + self.alpha * uncertainty)[0, 0]
            p.append(p_val)
        return np.argmax(p)

    def update(self, arm, reward, context):
        context = context.reshape(-1, 1)
        inv = self.A_inv[arm]
        u = inv @ context
        v_T = context.T @ inv
        denom = 1.0 + (context.T @ u)[0, 0]
        self.A_inv[arm] -= (u @ v_T) / denom
        self.b[arm] += reward * context

class EpsilonGreedy:
    def __init__(self, n_arms, epsilon=0.05):
        self.n_arms, self.epsilon = n_arms, epsilon
        self.counts, self.values = np.zeros(n_arms), np.zeros(n_arms)
    def select_arm(self, context=None):
        return np.random.randint(self.n_arms) if np.random.random() < self.epsilon else np.argmax(self.values)
    def update(self, arm, reward, context=None):
        self.counts[arm] += 1
        self.values[arm] += (reward - self.values[arm]) / self.counts[arm]

class UCB1:
    def __init__(self, n_arms):
        self.n_arms, self.counts, self.values, self.total_pulls = n_arms, np.zeros(n_arms), np.zeros(n_arms), 0
    def select_arm(self, context=None):
        self.total_pulls += 1
        if 0 in self.counts: return np.where(self.counts == 0)[0][0]
        return np.argmax(self.values + np.sqrt((2 * np.log(self.total_pulls)) / self.counts))
    def update(self, arm, reward, context=None):
        self.counts[arm] += 1
        self.values[arm] += (reward - self.values[arm]) / self.counts[arm]

# ==========================================
# 3. REPLAY EVALUATOR
# ==========================================

def run_oracle_replay(policy, df, context_matrix, oracle_probs, is_contextual=False, desc="Evaluating"):
    np.random.seed(42)
    indices = np.random.permutation(len(df))
    avg_reward_hist, cum_regret_hist, matches, total_reward, total_regret = [], [], 0, 0, 0
    
    for idx in tqdm(indices, desc=desc, leave=False):
        logged_arm, reward = df['arm'].iloc[idx], df['Reward'].iloc[idx]
        context = context_matrix[idx] if is_contextual else None
        chosen_arm = policy.select_arm(context) if is_contextual else policy.select_arm()
        
        if chosen_arm == logged_arm:
            if is_contextual: policy.update(chosen_arm, reward, context)
            else: policy.update(chosen_arm, reward)
            
            matches += 1
            total_reward += reward
            
            p0, p1 = oracle_probs[0][idx], oracle_probs[1][idx]
            best_p = max(p0, p1)
            chosen_p = p0 if chosen_arm == 0 else p1
            
            total_regret += (best_p - chosen_p)
            avg_reward_hist.append(total_reward / matches)
            cum_regret_hist.append(total_regret)
            
    return avg_reward_hist, cum_regret_hist, matches

# ==========================================
# 4. STEP 1: ALPHA HYPERPARAMETER SEARCH
# ==========================================
alphas = [0.1, 0.5, 1.0, 1.5, 2.0, 3.0, 4.0, 5.0]
alpha_results = {}

print("Step 1: Evaluating different Alphas for LinUCB...")
for a in alphas:
    res = run_oracle_replay(LinUCB(k_arms, n_features, alpha=a), df, X_context, oracle_probs, True, f"Alpha {a}")
    alpha_results[a] = res

# Find best alpha based on final CTR
best_alpha = max(alphas, key=lambda a: alpha_results[a][0][-1])

print("\n" + "="*45)
print(f"{'Alpha':<15} | {'Final CTR':<10} | {'Total Regret':<12}")
print("-" * 45)
for a in alphas:
    print(f"{a:<15} | {alpha_results[a][0][-1]:.5f} | {alpha_results[a][1][-1]:.2f}")
print("="*45)
print(f"Best Alpha found: {best_alpha}")

# ==========================================
# 5. STEP 2: FULL COMPARISON (using Best Alpha)
# ==========================================
print("\nStep 2: Evaluating Baselines and Final Comparison...")
res_eps = run_oracle_replay(EpsilonGreedy(k_arms), df, X_context, oracle_probs, False, "Eps-Greedy")
res_ucb = run_oracle_replay(UCB1(k_arms), df, X_context, oracle_probs, False, "UCB1")
res_lin_best = alpha_results[best_alpha]

print("\n" + "="*45)
print(f"{'Algorithm':<15} | {'Final CTR':<10} | {'Total Regret':<12}")
print("-" * 45)
print(f"{'Eps-Greedy':<15} | {res_eps[0][-1]:.5f} | {res_eps[1][-1]:.2f}")
print(f"{'UCB1':<15} | {res_ucb[0][-1]:.5f} | {res_ucb[1][-1]:.2f}")
print(f"{f'LinUCB (a={best_alpha})':<15} | {res_lin_best[0][-1]:.5f} | {res_lin_best[1][-1]:.2f}")
print("="*45)

# ==========================================
# 6. VISUALIZATION
# ==========================================
plt.figure(figsize=(20, 12))

# Subplot 1: Alpha Sweep CTR
plt.subplot(2, 3, 1)
for a in alphas:
    plt.plot(alpha_results[a][0], label=f'a={a}')
plt.title('Alpha Search: CTR Progression'); plt.ylim(0.1, 0.3)
plt.xlabel('Matches'); plt.ylabel('CTR'); plt.legend(); plt.grid(True, alpha=0.3)

# Subplot 2: Alpha Sweep Regret
plt.subplot(2, 3, 2)
for a in alphas:
    plt.plot(alpha_results[a][1], label=f'a={a}')
plt.title('Alpha Search: Cumulative Regret'); plt.xlabel('Matches'); plt.ylabel('Regret'); plt.legend(); plt.grid(True, alpha=0.3)

# Subplot 3: Final Comparison CTR
plt.subplot(2, 3, 4)
plt.plot(res_eps[0], label='Eps-Greedy')
plt.plot(res_ucb[0], label='UCB1')
plt.plot(res_lin_best[0], label=f'LinUCB (a={best_alpha})', color='red', linewidth=2)
plt.title('Final: CTR Progression'); plt.ylim(0.1, 0.3)
plt.xlabel('Matches'); plt.ylabel('CTR'); plt.legend(); plt.grid(True, alpha=0.3)

# Subplot 4: Final Comparison Regret
plt.subplot(2, 3, 5)
plt.plot(res_eps[1], label='Eps-Greedy')
plt.plot(res_ucb[1], label='UCB1')
plt.plot(res_lin_best[1], label=f'LinUCB (a={best_alpha})', color='red', linewidth=2)
plt.title('Final: Cumulative Regret'); plt.xlabel('Matches'); plt.ylabel('Regret'); plt.legend(); plt.grid(True, alpha=0.3)

# Subplot 5: Final Comparison Bar Chart
plt.subplot(2, 3, 6)
labels = ['Eps-Greedy', 'UCB1', f'LinUCB (a={best_alpha})']
values = [res_eps[0][-1], res_ucb[0][-1], res_lin_best[0][-1]]
bars = plt.bar(labels, values, color=['blue', 'orange', 'red'])
plt.title('Final Performance comparison'); plt.ylim(0.1, 0.3)
for b in bars: plt.text(b.get_x()+b.get_width()/2, b.get_height(), f'{b.get_height():.4f}', ha='center', va='bottom')

plt.tight_layout(); plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import xgboost as xgb
from sklearn.preprocessing import StandardScaler, OneHotEncoder, LabelEncoder
from sklearn.compose import ColumnTransformer
from tqdm.auto import tqdm
import os

# ==========================================
# 0. CONFIGURATION & ORACLE LOADING
# ==========================================
ORACLE_PATH = 'model_weights_depth_6.json'
df = mab_eval_df.copy()

# FILTER: Ensure banner_pos is strictly 0 or 1
df = df[df['banner_pos'].isin([0, 1])].reset_index(drop=True)

if os.path.exists(ORACLE_PATH):
    oracle_model = xgb.Booster()
    oracle_model.load_model(ORACLE_PATH)
    print(f"Successfully loaded Oracle: {ORACLE_PATH}")
else:
    raise FileNotFoundError(f"Oracle weight file not found at {ORACLE_PATH}.")

# ==========================================
# 1. DATA PREPARATION (Oracle vs MAB)
# ==========================================
df['Reward'] = df['click']
df['hour_of_day'] = df['hour'].apply(lambda x: int(str(x)[-2:]))

arm_names = [0, 1] 
k_arms = len(arm_names)
df['arm'] = df['banner_pos']

mab_categorical = ['C1', 'site_category', 'app_category', 'device_type', 'device_conn_type','C21']
mab_numeric = ['hour_of_day']

oracle_feature_order = ['hour_of_day', 'C1', 'banner_pos', 'site_category', 'app_category', 
                       'device_type', 'device_conn_type', 'C14', 'C17', 'C20', 'C21']

print("Encoding features for Oracle benchmark...")
oracle_input_df = df[oracle_feature_order].copy()
for col in oracle_feature_order:
    if col != 'hour_of_day':
        le = LabelEncoder()
        oracle_input_df[col] = le.fit_transform(oracle_input_df[col].astype(str))

print("Generating Oracle 'What-If' probabilities...")
oracle_probs = {}
for arm in arm_names:
    temp_df = oracle_input_df.copy()
    temp_df['banner_pos'] = arm  
    dmatrix = xgb.DMatrix(temp_df)
    oracle_probs[arm] = oracle_model.predict(dmatrix)

preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), mab_numeric),
        ('cat', OneHotEncoder(handle_unknown='ignore', sparse_output=False), mab_categorical)
    ])
X_context = preprocessor.fit_transform(df)
n_features = X_context.shape[1]
print(f"MAB Linear Feature Space: {n_features} dimensions")

# ==========================================
# 2. OPTIMIZED ALGORITHM DEFINITIONS
# ==========================================

class LinUCB:
    def __init__(self, n_arms, n_features, alpha=0.1):
        self.n_arms, self.n_features, self.alpha = n_arms, n_features, alpha
        self.A_inv = [np.eye(n_features) for _ in range(n_arms)]
        self.b = [np.zeros((n_features, 1)) for _ in range(n_arms)]

    def select_arm(self, context):
        context = context.reshape(-1, 1)
        p = []
        for a in range(self.n_arms):
            theta = self.A_inv[a] @ self.b[a]
            uncertainty = np.sqrt(context.T @ self.A_inv[a] @ context)
            p_val = (theta.T @ context + self.alpha * uncertainty)[0, 0]
            p.append(p_val)
        return np.argmax(p)

    def update(self, arm, reward, context):
        context = context.reshape(-1, 1)
        inv = self.A_inv[arm]
        u = inv @ context
        v_T = context.T @ inv
        denom = 1.0 + (context.T @ u)[0, 0]
        self.A_inv[arm] -= (u @ v_T) / denom
        self.b[arm] += reward * context

class EpsilonGreedy:
    def __init__(self, n_arms, epsilon=0.05):
        self.n_arms, self.epsilon = n_arms, epsilon
        self.counts, self.values = np.zeros(n_arms), np.zeros(n_arms)
    def select_arm(self, context=None):
        return np.random.randint(self.n_arms) if np.random.random() < self.epsilon else np.argmax(self.values)
    def update(self, arm, reward, context=None):
        self.counts[arm] += 1
        self.values[arm] += (reward - self.values[arm]) / self.counts[arm]

class UCB1:
    def __init__(self, n_arms):
        self.n_arms, self.counts, self.values, self.total_pulls = n_arms, np.zeros(n_arms), np.zeros(n_arms), 0
    def select_arm(self, context=None):
        self.total_pulls += 1
        if 0 in self.counts: return np.where(self.counts == 0)[0][0]
        return np.argmax(self.values + np.sqrt((2 * np.log(self.total_pulls)) / self.counts))
    def update(self, arm, reward, context=None):
        self.counts[arm] += 1
        self.values[arm] += (reward - self.values[arm]) / self.counts[arm]

# ==========================================
# 3. REPLAY EVALUATOR
# ==========================================

def run_oracle_replay(policy, df, context_matrix, oracle_probs, is_contextual=False, desc="Evaluating"):
    np.random.seed(42)
    indices = np.random.permutation(len(df))
    avg_reward_hist, cum_regret_hist, matches, total_reward, total_regret = [], [], 0, 0, 0
    
    for idx in tqdm(indices, desc=desc, leave=False):
        logged_arm, reward = df['arm'].iloc[idx], df['Reward'].iloc[idx]
        context = context_matrix[idx] if is_contextual else None
        chosen_arm = policy.select_arm(context) if is_contextual else policy.select_arm()
        
        if chosen_arm == logged_arm:
            if is_contextual: policy.update(chosen_arm, reward, context)
            else: policy.update(chosen_arm, reward)
            
            matches += 1
            total_reward += reward
            
            p0, p1 = oracle_probs[0][idx], oracle_probs[1][idx]
            best_p = max(p0, p1)
            chosen_p = p0 if chosen_arm == 0 else p1
            
            total_regret += (best_p - chosen_p)
            avg_reward_hist.append(total_reward / matches)
            cum_regret_hist.append(total_regret)
            
    return avg_reward_hist, cum_regret_hist, matches

# ==========================================
# 4. STEP 1: ALPHA HYPERPARAMETER SEARCH
# ==========================================
alphas = [0.1, 0.3, 0.5, 0.8, 1.0]
alpha_results = {}

print("Step 1: Evaluating different Alphas for LinUCB...")
for a in alphas:
    res = run_oracle_replay(LinUCB(k_arms, n_features, alpha=a), df, X_context, oracle_probs, True, f"Alpha {a}")
    alpha_results[a] = res

# Find best alpha based on final CTR
best_alpha = max(alphas, key=lambda a: alpha_results[a][0][-1])

print("\n" + "="*45)
print(f"{'Alpha':<15} | {'Final CTR':<10} | {'Total Regret':<12}")
print("-" * 45)
for a in alphas:
    print(f"{a:<15} | {alpha_results[a][0][-1]:.5f} | {alpha_results[a][1][-1]:.2f}")
print("="*45)
print(f"Best Alpha found: {best_alpha}")

# ==========================================
# 5. STEP 2: FULL COMPARISON (using Best Alpha)
# ==========================================
print("\nStep 2: Evaluating Baselines and Final Comparison...")
res_eps = run_oracle_replay(EpsilonGreedy(k_arms), df, X_context, oracle_probs, False, "Eps-Greedy")
res_ucb = run_oracle_replay(UCB1(k_arms), df, X_context, oracle_probs, False, "UCB1")
res_lin_best = alpha_results[best_alpha]

print("\n" + "="*45)
print(f"{'Algorithm':<15} | {'Final CTR':<10} | {'Total Regret':<12}")
print("-" * 45)
print(f"{'Eps-Greedy':<15} | {res_eps[0][-1]:.5f} | {res_eps[1][-1]:.2f}")
print(f"{'UCB1':<15} | {res_ucb[0][-1]:.5f} | {res_ucb[1][-1]:.2f}")
print(f"{f'LinUCB (a={best_alpha})':<15} | {res_lin_best[0][-1]:.5f} | {res_lin_best[1][-1]:.2f}")
print("="*45)

# ==========================================
# 6. VISUALIZATION
# ==========================================
plt.figure(figsize=(20, 12))

# Subplot 1: Alpha Sweep CTR
plt.subplot(2, 3, 1)
for a in alphas:
    plt.plot(alpha_results[a][0], label=f'a={a}')
plt.title('Alpha Search: CTR Progression'); plt.ylim(0.1, 0.3)
plt.xlabel('Matches'); plt.ylabel('CTR'); plt.legend(); plt.grid(True, alpha=0.3)

# Subplot 2: Alpha Sweep Regret
plt.subplot(2, 3, 2)
for a in alphas:
    plt.plot(alpha_results[a][1], label=f'a={a}')
plt.title('Alpha Search: Cumulative Regret'); plt.xlabel('Matches'); plt.ylabel('Regret'); plt.legend(); plt.grid(True, alpha=0.3)

# Subplot 3: Final Comparison CTR
plt.subplot(2, 3, 4)
plt.plot(res_eps[0], label='Eps-Greedy')
plt.plot(res_ucb[0], label='UCB1')
plt.plot(res_lin_best[0], label=f'LinUCB (a={best_alpha})', color='red', linewidth=2)
plt.title('Final: CTR Progression'); plt.ylim(0.1, 0.3)
plt.xlabel('Matches'); plt.ylabel('CTR'); plt.legend(); plt.grid(True, alpha=0.3)

# Subplot 4: Final Comparison Regret
plt.subplot(2, 3, 5)
plt.plot(res_eps[1], label='Eps-Greedy')
plt.plot(res_ucb[1], label='UCB1')
plt.plot(res_lin_best[1], label=f'LinUCB (a={best_alpha})', color='red', linewidth=2)
plt.title('Final: Cumulative Regret'); plt.xlabel('Matches'); plt.ylabel('Regret'); plt.legend(); plt.grid(True, alpha=0.3)

# Subplot 5: Final Comparison Bar Chart
plt.subplot(2, 3, 6)
labels = ['Eps-Greedy', 'UCB1', f'LinUCB (a={best_alpha})']
values = [res_eps[0][-1], res_ucb[0][-1], res_lin_best[0][-1]]
bars = plt.bar(labels, values, color=['blue', 'orange', 'red'])
plt.title('Final Performance comparison'); plt.ylim(0.1, 0.3)
for b in bars: plt.text(b.get_x()+b.get_width()/2, b.get_height(), f'{b.get_height():.4f}', ha='center', va='bottom')

plt.tight_layout(); plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import xgboost as xgb
from sklearn.preprocessing import StandardScaler, OneHotEncoder, LabelEncoder
from sklearn.compose import ColumnTransformer
from tqdm.auto import tqdm
import os

# ==========================================
# 0. CONFIGURATION & ORACLE LOADING
# ==========================================
ORACLE_PATH = 'model_weights_depth_6.json'
df = mab_eval_df.copy()

# FILTER: Ensure banner_pos is strictly 0 or 1
df = df[df['banner_pos'].isin([0, 1])].reset_index(drop=True)

if os.path.exists(ORACLE_PATH):
    oracle_model = xgb.Booster()
    oracle_model.load_model(ORACLE_PATH)
    print(f"Successfully loaded Oracle: {ORACLE_PATH}")
else:
    raise FileNotFoundError(f"Oracle weight file not found at {ORACLE_PATH}.")

# ==========================================
# 1. DATA PREPARATION (Oracle vs MAB)
# ==========================================
df['Reward'] = df['click']
df['hour_of_day'] = df['hour'].apply(lambda x: int(str(x)[-2:]))

arm_names = [0, 1] 
k_arms = len(arm_names)
df['arm'] = df['banner_pos']

mab_categorical = ['C1', 'site_category', 'app_category', 'device_type', 'device_conn_type','C21','C20','C14','C17']
mab_numeric = ['hour_of_day']

oracle_feature_order = ['hour_of_day', 'C1', 'banner_pos', 'site_category', 'app_category', 
                       'device_type', 'device_conn_type', 'C14', 'C17', 'C20', 'C21']

print("Encoding features for Oracle benchmark...")
oracle_input_df = df[oracle_feature_order].copy()
for col in oracle_feature_order:
    if col != 'hour_of_day':
        le = LabelEncoder()
        oracle_input_df[col] = le.fit_transform(oracle_input_df[col].astype(str))

print("Generating Oracle 'What-If' probabilities...")
oracle_probs = {}
for arm in arm_names:
    temp_df = oracle_input_df.copy()
    temp_df['banner_pos'] = arm  
    dmatrix = xgb.DMatrix(temp_df)
    oracle_probs[arm] = oracle_model.predict(dmatrix)

preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), mab_numeric),
        ('cat', OneHotEncoder(handle_unknown='ignore', sparse_output=False), mab_categorical)
    ])
X_context = preprocessor.fit_transform(df)
n_features = X_context.shape[1]
print(f"MAB Linear Feature Space: {n_features} dimensions")

# ==========================================
# 2. OPTIMIZED ALGORITHM DEFINITIONS
# ==========================================

class LinUCB:
    def __init__(self, n_arms, n_features, alpha=0.1):
        self.n_arms, self.n_features, self.alpha = n_arms, n_features, alpha
        self.A_inv = [np.eye(n_features) for _ in range(n_arms)]
        self.b = [np.zeros((n_features, 1)) for _ in range(n_arms)]

    def select_arm(self, context):
        context = context.reshape(-1, 1)
        p = []
        for a in range(self.n_arms):
            theta = self.A_inv[a] @ self.b[a]
            uncertainty = np.sqrt(context.T @ self.A_inv[a] @ context)
            p_val = (theta.T @ context + self.alpha * uncertainty)[0, 0]
            p.append(p_val)
        return np.argmax(p)

    def update(self, arm, reward, context):
        context = context.reshape(-1, 1)
        inv = self.A_inv[arm]
        u = inv @ context
        v_T = context.T @ inv
        denom = 1.0 + (context.T @ u)[0, 0]
        self.A_inv[arm] -= (u @ v_T) / denom
        self.b[arm] += reward * context

class EpsilonGreedy:
    def __init__(self, n_arms, epsilon=0.05):
        self.n_arms, self.epsilon = n_arms, epsilon
        self.counts, self.values = np.zeros(n_arms), np.zeros(n_arms)
    def select_arm(self, context=None):
        return np.random.randint(self.n_arms) if np.random.random() < self.epsilon else np.argmax(self.values)
    def update(self, arm, reward, context=None):
        self.counts[arm] += 1
        self.values[arm] += (reward - self.values[arm]) / self.counts[arm]

class UCB1:
    def __init__(self, n_arms):
        self.n_arms, self.counts, self.values, self.total_pulls = n_arms, np.zeros(n_arms), np.zeros(n_arms), 0
    def select_arm(self, context=None):
        self.total_pulls += 1
        if 0 in self.counts: return np.where(self.counts == 0)[0][0]
        return np.argmax(self.values + np.sqrt((2 * np.log(self.total_pulls)) / self.counts))
    def update(self, arm, reward, context=None):
        self.counts[arm] += 1
        self.values[arm] += (reward - self.values[arm]) / self.counts[arm]

# ==========================================
# 3. REPLAY EVALUATOR
# ==========================================

def run_oracle_replay(policy, df, context_matrix, oracle_probs, is_contextual=False, desc="Evaluating"):
    np.random.seed(42)
    indices = np.random.permutation(len(df))
    avg_reward_hist, cum_regret_hist, matches, total_reward, total_regret = [], [], 0, 0, 0
    
    for idx in tqdm(indices, desc=desc, leave=False):
        logged_arm, reward = df['arm'].iloc[idx], df['Reward'].iloc[idx]
        context = context_matrix[idx] if is_contextual else None
        chosen_arm = policy.select_arm(context) if is_contextual else policy.select_arm()
        
        if chosen_arm == logged_arm:
            if is_contextual: policy.update(chosen_arm, reward, context)
            else: policy.update(chosen_arm, reward)
            
            matches += 1
            total_reward += reward
            
            p0, p1 = oracle_probs[0][idx], oracle_probs[1][idx]
            best_p = max(p0, p1)
            chosen_p = p0 if chosen_arm == 0 else p1
            
            total_regret += (best_p - chosen_p)
            avg_reward_hist.append(total_reward / matches)
            cum_regret_hist.append(total_regret)
            
    return avg_reward_hist, cum_regret_hist, matches

# ==========================================
# 4. STEP 1: ALPHA HYPERPARAMETER SEARCH
# ==========================================
alphas = [0.1, 0.5, 1.0, 1.5]
alpha_results = {}

print("Step 1: Evaluating different Alphas for LinUCB...")
for a in alphas:
    res = run_oracle_replay(LinUCB(k_arms, n_features, alpha=a), df, X_context, oracle_probs, True, f"Alpha {a}")
    alpha_results[a] = res

# Find best alpha based on final CTR
best_alpha = max(alphas, key=lambda a: alpha_results[a][0][-1])

print("\n" + "="*45)
print(f"{'Alpha':<15} | {'Final CTR':<10} | {'Total Regret':<12}")
print("-" * 45)
for a in alphas:
    print(f"{a:<15} | {alpha_results[a][0][-1]:.5f} | {alpha_results[a][1][-1]:.2f}")
print("="*45)
print(f"Best Alpha found: {best_alpha}")

# ==========================================
# 5. STEP 2: FULL COMPARISON (using Best Alpha)
# ==========================================
print("\nStep 2: Evaluating Baselines and Final Comparison...")
res_eps = run_oracle_replay(EpsilonGreedy(k_arms), df, X_context, oracle_probs, False, "Eps-Greedy")
res_ucb = run_oracle_replay(UCB1(k_arms), df, X_context, oracle_probs, False, "UCB1")
res_lin_best = alpha_results[best_alpha]

print("\n" + "="*45)
print(f"{'Algorithm':<15} | {'Final CTR':<10} | {'Total Regret':<12}")
print("-" * 45)
print(f"{'Eps-Greedy':<15} | {res_eps[0][-1]:.5f} | {res_eps[1][-1]:.2f}")
print(f"{'UCB1':<15} | {res_ucb[0][-1]:.5f} | {res_ucb[1][-1]:.2f}")
print(f"{f'LinUCB (a={best_alpha})':<15} | {res_lin_best[0][-1]:.5f} | {res_lin_best[1][-1]:.2f}")
print("="*45)

# ==========================================
# 6. VISUALIZATION
# ==========================================
plt.figure(figsize=(20, 12))

# Subplot 1: Alpha Sweep CTR
plt.subplot(2, 3, 1)
for a in alphas:
    plt.plot(alpha_results[a][0], label=f'a={a}')
plt.title('Alpha Search: CTR Progression'); plt.ylim(0.1, 0.3)
plt.xlabel('Matches'); plt.ylabel('CTR'); plt.legend(); plt.grid(True, alpha=0.3)

# Subplot 2: Alpha Sweep Regret
plt.subplot(2, 3, 2)
for a in alphas:
    plt.plot(alpha_results[a][1], label=f'a={a}')
plt.title('Alpha Search: Cumulative Regret'); plt.xlabel('Matches'); plt.ylabel('Regret'); plt.legend(); plt.grid(True, alpha=0.3)

# Subplot 3: Final Comparison CTR
plt.subplot(2, 3, 4)
plt.plot(res_eps[0], label='Eps-Greedy')
plt.plot(res_ucb[0], label='UCB1')
plt.plot(res_lin_best[0], label=f'LinUCB (a={best_alpha})', color='red', linewidth=2)
plt.title('Final: CTR Progression'); plt.ylim(0.1, 0.3)
plt.xlabel('Matches'); plt.ylabel('CTR'); plt.legend(); plt.grid(True, alpha=0.3)

# Subplot 4: Final Comparison Regret
plt.subplot(2, 3, 5)
plt.plot(res_eps[1], label='Eps-Greedy')
plt.plot(res_ucb[1], label='UCB1')
plt.plot(res_lin_best[1], label=f'LinUCB (a={best_alpha})', color='red', linewidth=2)
plt.title('Final: Cumulative Regret'); plt.xlabel('Matches'); plt.ylabel('Regret'); plt.legend(); plt.grid(True, alpha=0.3)

# Subplot 5: Final Comparison Bar Chart
plt.subplot(2, 3, 6)
labels = ['Eps-Greedy', 'UCB1', f'LinUCB (a={best_alpha})']
values = [res_eps[0][-1], res_ucb[0][-1], res_lin_best[0][-1]]
bars = plt.bar(labels, values, color=['blue', 'orange', 'red'])
plt.title('Final Performance comparison'); plt.ylim(0.1, 0.3)
for b in bars: plt.text(b.get_x()+b.get_width()/2, b.get_height(), f'{b.get_height():.4f}', ha='center', va='bottom')

plt.tight_layout(); plt.show()
```
